## дәҢгҖҒ и§ЈеҶізҡ„ж ёеҝғй—®йўҳ


 &emsp;&emsp;дј з»ҹзҡ„ Agent ејҖеҸ‘йҖҡеёёиҝҗиЎҢдёҖдёӘз®ҖеҚ•зҡ„еҫӘзҺҜпјҡ**жҖқиҖғвҶ’и°ғз”Ёе·Ҙе…·вҶ’и§ӮеҜҹвҶ’йҮҚеӨҚ**гҖӮиҝҷз§ҚжЁЎејҸеңЁеӨ„зҗҶеӨҡе°Ҹж—¶жҲ–еӨҡеӨ©зҡ„д»»еҠЎж—¶пјҢе®№жҳ“йҒҮеҲ°д»ҘдёӢ"жө…еұӮйҷ·йҳұ"пјҲShallow Agent Problemпјү

1. **и§„еҲ’иғҪеҠӣзјәеӨұ**пјҡеҺҹз”ҹ Agent еҖҫеҗ‘дәҺвҖңиө°дёҖжӯҘзңӢдёҖжӯҘвҖқпјҢзјәд№Ҹе…ЁеұҖи§Ҷи§’зҡ„д»»еҠЎжӢҶи§ЈпјҢе®№жҳ“еңЁеӨҡжӯҘд»»еҠЎдёӯиҝ·еӨұж–№еҗ‘гҖӮ

2. **йҒ—еҝҳдёҺж··д№ұ**пјҡеңЁжү§иЎҢи¶…иҝҮ 10-20 жӯҘзҡ„й•ҝд»»еҠЎж—¶пјҢз”ұдәҺ Context WindowпјҲдёҠдёӢж–ҮзӘ—еҸЈпјүйҷҗеҲ¶пјҢдј з»ҹ Agent е®№жҳ“еҝҳи®°еҲқе§Ӣзӣ®ж ҮжҲ–йҷ·е…Ҙжӯ»еҫӘзҺҜгҖӮ

3. **зҺҜеўғдәӨдә’еӣ°йҡҫ**пјҡж–Үд»¶зі»з»ҹж“ҚдҪңгҖҒд»Јз Ғжү§иЎҢзҺҜеўғпјҲжІҷз®ұпјүзҡ„й…ҚзҪ®е’Ңе®үе…Ёз®ЎзҗҶеӨҚжқӮгҖӮ

4. **дёҠдёӢж–ҮжұЎжҹ“**пјҡжүҖжңүе·Ҙе…·иҝ”еӣһз»“жһңйғҪе Ҷз§ҜеңЁдёҖдёӘеҜ№иҜқеҺҶеҸІдёӯпјҢеҜјиҮҙеҷӘеЈ°иҝҮеӨ§гҖӮ

5. **еҚҸдҪңзј–жҺ’еӨҚжқӮ**пјҡеӨҡжҷәиғҪдҪ“пјҲMulti-Agentпјүд№Ӣй—ҙзҡ„д»»еҠЎеҲҶеҸ‘е’ҢдёҠдёӢж–Үйҡ”зҰ»йҡҫд»Ҙе®һзҺ°гҖӮ

## дёүгҖҒ еә”з”ЁеңәжҷҜ

````python
from deepagents import create_deep_agent
from langchain_tavily import TavilySearch
from langgraph.checkpoint.memory import InMemorySaver

# 1. еҲқе§ӢеҢ– Tavily жҗңзҙўе·Ҙе…·
tavily = TavilySearch(max_results=3)



# 2. зј–еҶҷзі»з»ҹжҸҗзӨәиҜҚ
research_instructions = """
жӮЁжҳҜдёҖдҪҚиө„ж·ұзҡ„з ”з©¶дәәе‘ҳгҖӮжӮЁзҡ„е·ҘдҪңжҳҜиҝӣиЎҢж·ұе…Ҙзҡ„з ”з©¶пјҢз„¶еҗҺж’°еҶҷдёҖд»ҪзІҫзҫҺзҡ„жҠҘе‘ҠгҖӮ
жӮЁеҸҜд»ҘйҖҡиҝҮдә’иҒ”зҪ‘жҗңзҙўеј•ж“ҺдҪңдёәдё»иҰҒзҡ„дҝЎжҒҜж”¶йӣҶе·Ҙе…·гҖӮ
## еҸҜз”Ёе·Ҙе…·
### `дә’иҒ”зҪ‘жҗңзҙў`
дҪҝз”ЁжӯӨеҠҹиғҪй’ҲеҜ№з»ҷе®ҡзҡ„жҹҘиҜўиҝӣиЎҢдә’иҒ”зҪ‘жҗңзҙўгҖӮжӮЁеҸҜд»ҘжҢҮе®ҡиҰҒиҝ”еӣһзҡ„жңҖеӨ§з»“жһңж•°йҮҸгҖҒдё»йўҳд»ҘеҸҠжҳҜеҗҰеҢ…еҗ«еҺҹе§ӢеҶ…е®№гҖӮ
### `еҶҷе…Ҙжң¬ең°ж–Үд»¶`
дҪҝз”ЁжӯӨеҠҹиғҪе°Ҷз ”з©¶жҠҘе‘ҠдҝқеӯҳеҲ°жң¬ең°ж–Үд»¶гҖӮеҪ“жӮЁе®ҢжҲҗз ”з©¶е№¶з”ҹжҲҗжҠҘе‘ҠеҗҺпјҢиҜ·дҪҝз”ЁжӯӨе·Ҙе…·е°Ҷе®Ңж•ҙзҡ„жҠҘе‘ҠеҶ…е®№дҝқеӯҳеҲ°ж–Ү
д»¶дёӯгҖӮ
- ж–Үд»¶и·Ҝеҫ„е»әи®®дҪҝз”Ё .md ж јејҸпјҲMarkdownпјүпјҢдҫӢеҰӮ "research_report.md" жҲ– "./reports/жҠҘе‘ҠеҗҚ
з§°.md"
- иҜ·зЎ®дҝқжҠҘе‘ҠеҶ…е®№е®Ңж•ҙгҖҒз»“жһ„жё…жҷ°пјҢеҢ…еҗ«жүҖжңүз« иҠӮе’Ңеј•з”ЁжқҘжәҗ
## е·ҘдҪңжөҒзЁӢ
еңЁиҝӣиЎҢз ”з©¶ж—¶пјҡ
1. йҰ–е…Ҳе°Ҷз ”з©¶д»»еҠЎеҲҶи§Јдёәжё…жҷ°зҡ„жӯҘйӘӨ
2. дҪҝз”Ёдә’иҒ”зҪ‘жҗңзҙўжқҘж”¶йӣҶе…Ёйқўзҡ„дҝЎжҒҜ
3. е°ҶдҝЎжҒҜж•ҙеҗҲжҲҗдёҖд»Ҫз»“жһ„жё…жҷ°зҡ„жҠҘе‘Ҡ
4. **йҮҚиҰҒ**пјҡе®ҢжҲҗжҠҘе‘ҠеҗҺпјҢеҠЎеҝ…дҪҝз”Ё `еҶҷе…Ҙжң¬ең°ж–Үд»¶` е·Ҙе…·е°Ҷе®Ңж•ҙжҠҘе‘ҠдҝқеӯҳеҲ°жң¬ең°ж–Үд»¶
5. еҠЎеҝ…еј•з”ЁдҪ зҡ„иө„ж–ҷжқҘжәҗ
**жіЁж„Ҹ**пјҡиҜ·зЎ®дҝқеңЁе®ҢжҲҗз ”з©¶еҗҺпјҢе°Ҷе®Ңж•ҙзҡ„жҠҘе‘ҠеҶ…е®№дҝқеӯҳеҲ°ж–Үд»¶дёӯпјҢиҝҷж ·з”ЁжҲ·еҸҜд»Ҙж–№дҫҝең°жҹҘзңӢе’ҢдҝқеӯҳжҠҘе‘ҠгҖӮ
"""

# 2. еҲӣе»ә DeepAgents жҷәиғҪдҪ“
agent = create_deep_agent(
    name="DeepAgents_Agent",       # жҷәиғҪдҪ“еҗҚз§°
    tools=[tavily],                # еҸҜи°ғз”Ёе·Ҙе…·Tool
    model=model,                   # жЁЎеһӢModel
    system_prompt=research_instructions,  # зі»з»ҹжҸҗзӨәиҜҚ
    checkpointer=InMemorySaver(),  # жЈҖжҹҘзӮ№CheckpointerпјҢеҶ…еӯҳжЈҖжҹҘзӮ№
)

# 3. й…ҚзҪ®зәҝзЁӢ ID
config = {"configurable": {"thread_id": "1"}}

result = agent.invoke({"messages": [{"role": "user", "content": "её®жҲ‘жҹҘиҜўдёҖдёӢжңүе…іdeepagentsжЎҶжһ¶зҡ„жңҖж–°еҠЁжҖҒ"}]}, config=config)
````

````python
# еҜје…ҘRichеә“пјҢз”ЁдәҺзҫҺеҢ–д»Јз Ғ
from rich.console import Console
from rich.table import Table
from rich.panel import Panel
RICH_AVAILABLE = True
console = Console()

def print_agent_tools(agent):
    """
    жү“еҚ° Agent дёӯеҠ иҪҪзҡ„жүҖжңүе·Ҙе…·
    еҢ…жӢ¬з”ЁжҲ·иҮӘе®ҡд№үе·Ҙе…·гҖҒж–Үд»¶зі»з»ҹе·Ҙе…·гҖҒзі»з»ҹе·Ҙе…·зӯү
    """
    # иҺ·еҸ– agent зҡ„ nodes (LangGraph зҡ„иҠӮзӮ№)
    if hasattr(agent, 'nodes') and 'tools' in agent.nodes:
        tools_node = agent.nodes['tools']

        # tools_node жҳҜ PregelNodeпјҢзңҹжӯЈзҡ„ ToolNode еңЁ bound еұһжҖ§дёӯ
        if hasattr(tools_node, 'bound'):
            tool_node = tools_node.bound

            # д»Һ ToolNode иҺ·еҸ–е·Ҙе…·
            if hasattr(tool_node, 'tools_by_name'):
                tools = tool_node.tools_by_name

                # еҲҶзұ»е·Ҙе…·
                user_tools = []
                filesystem_tools = []
                system_tools = []

                for tool_name, tool in tools.items():
                    tool_info = {
                        'name': tool_name,
                        'description': getattr(tool, 'description', 'ж— жҸҸиҝ°')
                    }

                    # еҲҶзұ»
                    if tool_name in ['ls', 'read_file', 'write_file', 'edit_file', 'glob', 'grep', 'execute']:
                        filesystem_tools.append(tool_info)
                    elif tool_name in ['write_todos', 'task']:
                        system_tools.append(tool_info)
                    else:
                        user_tools.append(tool_info)

                # жү“еҚ°еҠ иҪҪе·Ҙе…·зҡ„иҫ“еҮә
                _print_tools_rich(user_tools, filesystem_tools, system_tools)

            else:
                print("ж— жі•иҺ·еҸ–е·Ҙе…·еҲ—иЎЁ (tools_by_name дёҚеӯҳеңЁ)")
        else:
            print("ж— жі•иҺ·еҸ–е·Ҙе…·еҲ—иЎЁ (bound еұһжҖ§дёҚеӯҳеңЁ)")
    else:
        print("ж— жі•иҺ·еҸ–е·Ҙе…·еҲ—иЎЁ (nodes з»“жһ„дёҚз¬ҰеҗҲйў„жңҹ)")

def _print_tools_rich(user_tools, filesystem_tools, system_tools):
    """дҪҝз”Ё Rich еә“зҫҺеҢ–жү“еҚ°е·Ҙе…·еҲ—иЎЁ"""
    console.print()

    # еҲӣе»әиЎЁж ј
    table = Table(title="Agent еҠ иҪҪзҡ„е·Ҙе…·еҲ—иЎЁ", show_header=True, header_style="bold magenta")
    table.add_column("зұ»еҲ«", style="cyan", width=20)
    table.add_column("е·Ҙе…·еҗҚз§°", style="green", width=20)
    table.add_column("жҸҸиҝ°", style="white", width=60)

    # ж·»еҠ з”ЁжҲ·е·Ҙе…·
    for i, tool in enumerate(user_tools):
        category = "з”ЁжҲ·е·Ҙе…·" if i == 0 else ""
        desc = tool['description'][:80] + "..." if len(tool['description']) > 80 else tool['description']
        table.add_row(category, tool['name'], desc)

    # ж·»еҠ ж–Үд»¶зі»з»ҹе·Ҙе…·
    for i, tool in enumerate(filesystem_tools):
        category = "ж–Үд»¶зі»з»ҹе·Ҙе…·" if i == 0 else ""
        desc = tool['description'][:80] + "..." if len(tool['description']) > 80 else tool['description']
        table.add_row(category, tool['name'], desc)

    # ж·»еҠ зі»з»ҹе·Ҙе…·
    for i, tool in enumerate(system_tools):
        category = "зі»з»ҹе·Ҙе…·" if i == 0 else ""
        desc = tool['description'][:80] + "..." if len(tool['description']) > 80 else tool['description']
        table.add_row(category, tool['name'], desc)

    console.print(table)

    # жү“еҚ°з»ҹи®Ў
    total = len(user_tools) + len(filesystem_tools) + len(system_tools)
    console.print(Panel(
        f"[bold green]е…ұи®Ў {total} дёӘе·Ҙе…·[/bold green]\n\n"
        f"вҖў з”ЁжҲ·е·Ҙе…·: {len(user_tools)} дёӘ\n"
        f"вҖў ж–Үд»¶зі»з»ҹе·Ҙе…·: {len(filesystem_tools)} дёӘ\n"
        f"вҖў зі»з»ҹе·Ҙе…·: {len(system_tools)} дёӘ",
        title="з»ҹи®ЎдҝЎжҒҜ",
        border_style="green"
    ))
    console.print()
````

# дәҢгҖҒ и§„еҲ’е·Ҙе…· (Planning System / Todo List)


- **з»„д»¶**пјҡ`TodoListMiddleware`

- **е·Ҙе…·еҗҚ**пјҡ`write_todos`

- **еҠҹиғҪ**пјҡAgent еңЁиЎҢеҠЁеүҚе…Ҳз”ҹжҲҗ Markdown ж јејҸзҡ„еҫ…еҠһдәӢйЎ№еҲ—иЎЁ (Todo List)пјҢе№¶еңЁжү§иЎҢиҝҮзЁӢдёӯжӣҙж–°зҠ¶жҖҒпјҲе®ҢжҲҗ/иҝӣиЎҢдёӯпјүгҖӮ

- **е·ҘдҪңжөҒ**пјҡ

    1. Agent жҺҘж”¶еӨҚжқӮд»»еҠЎпјҲз®ҖеҚ•зҹӯжңҹзҡ„д»»еҠЎдёҚдјҡи§ҰеҸ‘todolistпјүгҖӮ

    2. и°ғз”Ё `write_todos` е°Ҷд»»еҠЎжӢҶи§ЈдёәеӯҗжӯҘйӘӨ (Pending)гҖӮ

    3. жҜҸе®ҢжҲҗдёҖжӯҘпјҢжӣҙж–°зҠ¶жҖҒдёә (Completed)гҖӮ

    4. **иҮӘжҲ‘еҸҚжҖқ**пјҡеңЁжҜҸдёҖжӯҘиЎҢеҠЁеүҚпјҢAgent йғҪдјҡзңӢеҲ°еҪ“еүҚзҡ„ Todo ListпјҢд»ҺиҖҢйҒҝе…Қиҝ·еӨұж–№еҗ‘гҖӮиҝҷејәеҲ¶жЁЎеһӢиҝӣиЎҢ"жҖқз»ҙй“ҫ"зҡ„жҳҫжҖ§еҢ–з®ЎзҗҶгҖӮ

# дёүгҖҒ еӯҗд»ЈзҗҶ (Sub-Agent Delegation)


- **з»„д»¶**пјҡ`SubAgentMiddleware`

- **е·Ҙе…·еҗҚ**пјҡ`task` (delegate_task)

- **ж ёеҝғжҰӮеҝө**пјҡ

  - **д»»еҠЎйҡ”зҰ»**пјҡжҜҸдёӘеӯҗд»ЈзҗҶжңүзӢ¬з«Ӣзҡ„дёҠдёӢж–ҮзӘ—еҸЈгҖӮ

  - **е№¶иЎҢжү§иЎҢ**пјҡж”ҜжҢҒеҗҢж—¶еҗҜеҠЁеӨҡдёӘеӯҗд»ЈзҗҶгҖӮ

  - **з»“жһңиҒҡеҗҲ**пјҡжҷәиғҪж•ҙеҗҲеӨҡдёӘеӯҗд»ЈзҗҶзҡ„иҫ“еҮәгҖӮ

- **жңәеҲ¶**пјҡ

  - еҪ“д»»еҠЎиҝҮдәҺе…·дҪ“пјҲеҰӮ"зҲ¬еҸ–е№¶еҲҶжһҗиҝҷзҜҮй•ҝи®әж–Ү"пјүж—¶пјҢдё» Agent дјҡз”ҹжҲҗдёҖдёӘйҡ”зҰ»зҺҜеўғзҡ„еӯҗ Agent еҺ»жү§иЎҢгҖӮ

  - еӯҗ Agent еҗҜеҠЁж—¶пјҢеҸӘз»§жүҝеҝ…иҰҒзҡ„зҺҜеўғй…ҚзҪ®пјҢдҪҶ**жӢҘжңүе…Ёж–°зҡ„гҖҒз©әзҷҪзҡ„ж¶ҲжҒҜеҺҶеҸІ**гҖӮ

  - еӯҗ Agent жү§иЎҢе®ҢжҜ•еҗҺеҸӘиҝ”еӣһдёҖдёӘжҖ»з»“жҖ§зҡ„з»“жһңгҖӮиҝҷдҝқиҜҒдәҶдё» Agent зҡ„ж—¶й—ҙзәҝпјҲContextпјүдҝқжҢҒе№ІеҮҖпјҢжһҒеӨ§ең°иҠӮзңҒдәҶ TokenгҖӮ


### иҮӘеҠЁи§ҰеҸ‘й»ҳи®Өзҡ„ SubAgentMiddleware

````python
import asyncio
import os
from dotenv import load_dotenv
from rich.console import Console
from rich.panel import Panel
from rich.tree import Tree
from deepagents import create_deep_agent
from langchain_openai import ChatOpenAI
from langchain_community.tools import TavilySearchResults

from langchain_core.messages import ToolMessage, BaseMessage

# еҠ иҪҪзҺҜеўғеҸҳйҮҸ
load_dotenv(override=True)

# й…ҚзҪ® Rich Console
console = Console()

async def run_auto_subagent_demo():
    """
    жј”зӨәпјҡдёҚжҳҫејҸдј е…Ҙ subagents еҸӮж•°пјҢиҮӘеҠЁи§ҰеҸ‘й»ҳи®Өзҡ„ SubAgentMiddleware
    """
    console.print(Panel.fit("[bold magenta]DeepAgents иҮӘеҠЁ SubAgent дёӯй—ҙд»¶жј”зӨә[/bold magenta]", border_style="magenta"))
    console.print("[dim]жң¬жј”зӨәйӘҢиҜҒпјҡеҚідҪҝдёҚдј е…Ҙ subagents еҸӮж•°пјҢAgent й»ҳи®Өд№ҹдјҡеҗҜз”Ё 'general-purpose' еӯҗ AgentгҖӮ[/dim]")

    # 1. еҲқе§ӢеҢ–жЁЎеһӢ
    # дҪҝз”Ё GPT-4o д»ҘзЎ®дҝқеҜ№еӨҚжқӮжҢҮд»Өзҡ„зҗҶи§Је’Ңе·Ҙе…·и°ғз”Ёзҡ„еҮҶзЎ®жҖ§
    llm = ChatOpenAI(model="gpt-4o", temperature=0)

    # 2. е®ҡд№үе·Ҙе…·
    # з»ҷ Agent дёҖдәӣе·Ҙе…·пјҢд»Ҙдҫҝеӯҗ Agent д№ҹжңүе·Ҙе…·еҸҜз”Ё
    tools = [TavilySearchResults(max_results=2)]

    # 3. еҲӣе»ә Agent (дёҚдј е…Ҙ subagents)
    # create_deep_agent й»ҳи®ӨдјҡеҠ иҪҪ SubAgentMiddleware(general_purpose_agent=True)
    # иҝҷж„Ҹе‘ізқҖ Agent дјҡиҮӘеҠЁиҺ·еҫ—дёҖдёӘеҗҚдёә 'task' зҡ„е·Ҙе…·пјҢеҸҜд»Ҙи°ғз”Ё 'general-purpose' еӯҗ Agent
    console.print("[bold cyan]жӯЈеңЁеҲӣе»ә Agent (subagents=None)...[/bold cyan]")

    agent = create_deep_agent(
        model=llm,
        tools=tools,
        # subagents=[], # ж•…ж„ҸдёҚдј жҲ–дј з©ә
        system_prompt="""дҪ жҳҜдёҖдёӘиғҪеӨҹй«ҳж•ҲеӨ„зҗҶе№¶еҸ‘д»»еҠЎзҡ„жҷәиғҪеҠ©жүӢгҖӮ
                        еҜ№дәҺеҢ…еҗ«еӨҡдёӘзӢ¬з«ӢйғЁеҲҶзҡ„еӨҚжқӮд»»еҠЎпјҢдҪ еҝ…йЎ»дҪҝз”Ё 'task' е·Ҙе…·жқҘеҲӣе»ә 'general-purpose' еӯҗ Agent иҝӣиЎҢеӨ„зҗҶгҖӮ
                        дёҚиҰҒиҮӘе·ұеңЁдё»зәҝзЁӢдёӯдёІиЎҢжү§иЎҢжүҖжңүж“ҚдҪңгҖӮеҲ©з”Ёеӯҗ Agent жқҘйҡ”зҰ»дёҠдёӢж–Үе№¶жҸҗй«ҳж•ҲзҺҮгҖӮ"""
    )

    # 4. е®ҡд№үдёҖдёӘйҖӮеҗҲе№¶иЎҢ/йҡ”зҰ»зҡ„д»»еҠЎ
    task = """иҜ·еҗҢж—¶и°ғз ”д»ҘдёӢдёӨдёӘе®Ңе…ЁдёҚеҗҢзҡ„дё»йўҳпјҢе№¶еҲҶеҲ«з»ҷеҮәз®ҖзҹӯжҖ»з»“пјҡ
            1. Python иҜӯиЁҖзҡ„еҺҶеҸІиө·жәҗгҖӮ
            2. Rust иҜӯиЁҖзҡ„еҶ…еӯҳе®үе…ЁжңәеҲ¶гҖӮ
            иҜ·еҠЎеҝ…дҪҝз”Ёеӯҗ Agent еҲҶеҲ«еӨ„зҗҶиҝҷдёӨдёӘд»»еҠЎгҖӮ"""

    console.print(f"\n[bold green]з”ЁжҲ·д»»еҠЎ:[/bold green] {task}\n")

    # 5. иҝҗиЎҢе№¶еҸҜи§ҶеҢ–
    step = 0

    # жЈҖжҹҘ API Key
    if not os.getenv("OPENAI_API_KEY"):
        console.print("[bold red]вқҢ й”ҷиҜҜ: жңӘжүҫеҲ° OPENAI_API_KEYпјҢиҜ·жЈҖжҹҘ .env ж–Үд»¶[/bold red]")
        return

    console.print("[dim]ејҖе§ӢжөҒејҸиҫ“еҮә...[/dim]")
    try:
        async for event in agent.astream({"messages": [("user", task)]}):
            step += 1
            # йҒҚеҺҶжүҖжңүиҠӮзӮ№иҫ“еҮә (e.g., 'agent', 'tools')
            for node_name, node_data in event.items():
                if node_data is None:
                    continue

                if "messages" in node_data:
                    msgs = node_data["messages"]
                    # зЎ®дҝқжҳҜеҲ—иЎЁ
                    if not isinstance(msgs, list):
                        msgs = [msgs]

                    for msg in msgs:
                        # 0. иҝҮж»Өйқһж¶ҲжҒҜеҜ№иұЎ
                        if not isinstance(msg, BaseMessage):
                            continue

                        # 1. жЈҖжөӢе·Ҙе…·и°ғз”Ё (жңҹжңӣзңӢеҲ° 'task' е·Ҙе…·)
                        if hasattr(msg, "tool_calls") and msg.tool_calls:
                            tree = Tree(f"[bold yellow]Step {step}: еҶізӯ–дёҺи°ғз”Ё (Node: {node_name})[/bold yellow]")
                            for tc in msg.tool_calls:
                                tool_name = tc['name']
                                tool_args = tc['args']

                                if tool_name == "task":
                                    # йӘҢиҜҒжҲҗеҠҹпјҒ
                                    branch = tree.add(f"[bold red]рҹҡҖ и§ҰеҸ‘ 'task' е·Ҙе…· (Sub-Agent)[/bold red]")
                                    branch.add(f"[cyan]еӯҗ Agent зұ»еһӢ:[/cyan] {tool_args.get('subagent_type')}")
                                    branch.add(f"[cyan]д»»еҠЎжҢҮд»Ө:[/cyan] {tool_args.get('description')}")
                                else:
                                    tree.add(f"[blue]жҷ®йҖҡе·Ҙе…·и°ғз”Ё:[/blue] {tool_name}")

                            console.print(tree)

                        # 2. жЈҖжөӢе·Ҙе…·иҫ“еҮә (Sub-Agent зҡ„иҝ”еӣһз»“жһң)
                        elif isinstance(msg, ToolMessage):
                            if msg.name == "task":
                                # Sub-Agent е®ҢжҲҗд»»еҠЎиҝ”еӣһ
                                panel = Panel(
                                    msg.content,
                                    title=f"[bold magenta]Sub-Agent е®ҢжҲҗд»»еҠЎ (Node: {node_name})[/bold magenta]",
                                    border_style="magenta"
                                )
                                console.print(panel)
                            else:
                                console.print(f"[dim]Tool Output ({msg.name}): {msg.content[:100]}...[/dim]")

                        # 3. жЈҖжөӢ AI жңҖз»ҲеӣһеӨҚ
                        elif msg.content and not msg.tool_calls:
                            title = f"[bold green]Agent еӣһеӨҚ (Node: {node_name})[/bold green]"
                            console.print(Panel(msg.content, title=title, border_style="green"))

    except Exception as e:
        console.print(f"[bold red]вқҢ иҝҗиЎҢж—¶й”ҷиҜҜ: {e}[/bold red]")

    console.print("\n[bold magenta]жј”зӨәз»“жқҹ[/bold magenta]")

if __name__ == "__main__":
    await run_auto_subagent_demo()
````

````python
import asyncio
import os
from dotenv import load_dotenv
from rich.console import Console
from rich.panel import Panel
from rich.tree import Tree
from deepagents import create_deep_agent
from langchain_deepseek import ChatDeepSeek
from langchain_mcp_adapters.client import MultiServerMCPClient
from langchain_community.tools import TavilySearchResults

from langchain_core.messages import ToolMessage, BaseMessage

# еҠ иҪҪзҺҜеўғеҸҳйҮҸ
load_dotenv(override=True)

# 1. й…ҚзҪ® Rich Console
console = Console()

# 2. й…ҚзҪ® Context7 MCP (иҝһжҺҘе®ҳж–№ж–ҮжЎЈ)
async def setup_mcp_tools():
    console.print("[dim]жӯЈеңЁиҝһжҺҘ Context7 MCP жңҚеҠЎеҷЁ...[/dim]")
    # жЈҖжҹҘ node зҺҜеўғ
    try:
        client = MultiServerMCPClient({
            "context7": {
                "transport": "stdio",
                "command": "npx",
                "args": ["-y", "@upstash/context7-mcp@latest"],
            }
        })
        # иҺ·еҸ–е·Ҙе…·
        tools = await client.get_tools()
        console.print(f"[green]жҲҗеҠҹеҠ иҪҪ {len(tools)} дёӘ MCP е·Ҙе…·[/green]")
        return client, tools
    except Exception as e:
        console.print(f"[red]иҝһжҺҘ MCP еӨұиҙҘ: {e}[/red]")
        console.print("[yellow]е°ҶдҪҝз”ЁжЁЎжӢҹе·Ҙе…·з»§з»ӯ...[/yellow]")
        return None, []

# 3. е®ҡд№үеӯҗ Agent й…ҚзҪ®
def get_subagents_config(mcp_tools):
    # еӯҗ Agent 1: е®ҳж–№ж–ҮжЎЈдё“е®¶
    doc_tools = mcp_tools if mcp_tools else [TavilySearchResults(max_results=3)]

    docs_researcher = {
        "name": "DocsResearcher",
        "description": "иҙҹиҙЈжҹҘйҳ…е®ҳж–№ж–ҮжЎЈе’ҢжҠҖжңҜи§„иҢғзҡ„дё“е®¶ AgentгҖӮ",
        "system_prompt": "дҪ жҳҜдёҖеҗҚдё“й—ЁжҹҘйҳ…е®ҳж–№ж–ҮжЎЈзҡ„жҠҖжңҜдё“е®¶гҖӮиҜ·дҪҝз”Ёе·Ҙе…·иҺ·еҸ–еҮҶзЎ®зҡ„жҠҖжңҜз»ҶиҠӮгҖӮдёҚиҰҒзҢңжөӢгҖӮ",
        "tools": doc_tools,
        "model": "deepseek-chat"
    }

    # еӯҗ Agent 2: зӨҫеҢәз”ҹжҖҒдё“е®¶
    community_researcher = {
        "name": "CommunityResearcher",
        "description": "иҙҹиҙЈжҗңзҙўзӨҫеҢәеҚҡе®ўгҖҒж•ҷзЁӢе’ҢжңҖдҪіе®һи·өзҡ„дё“е®¶ AgentгҖӮ",
        "system_prompt": "дҪ жҳҜдёҖеҗҚе…іжіЁзӨҫеҢәеҠЁжҖҒзҡ„ејҖеҸ‘иҖ…гҖӮиҜ·жҗңзҙўеҚҡе®ўгҖҒи®әеқӣе’Ң GitHub и®Ёи®әгҖӮ",
        "tools": [TavilySearchResults(max_results=3)],
        "model": "deepseek-chat"
    }

    return [docs_researcher, community_researcher]

# 4. дё»иҝҗиЎҢйҖ»иҫ‘
async def run_parallel_demo():
    console.print(Panel.fit("[bold blue]DeepAgents е№¶иЎҢеӯҗ Agent жј”зӨә[/bold blue]", border_style="blue"))

    # еҲқе§ӢеҢ– MCP
    mcp_client, mcp_tools = await setup_mcp_tools()

    # иҺ·еҸ–еӯҗ Agent й…ҚзҪ®
    subagents = get_subagents_config(mcp_tools)

    # еҲӣе»әдё» Agent
    llm = ChatDeepSeek(model="deepseek-chat", temperature=0)

    agent = create_deep_agent(
        model=llm,
        tools=[],
        subagents=subagents,        # дј е…Ҙеӯҗ Agent й…ҚзҪ®
        system_prompt="""дҪ жҳҜдёҖеҗҚжҠҖжңҜжҖ»зӣ‘гҖӮдҪ зҡ„д»»еҠЎжҳҜеҚҸи°ғ DocsResearcher е’Ң CommunityResearcher е®ҢжҲҗи°ғз ”д»»еҠЎгҖӮ
                        иҜ·ж №жҚ®з”ЁжҲ·йңҖжұӮпјҢе°Ҷд»»еҠЎжӢҶи§Је№¶еҲҶеҸ‘з»ҷиҝҷдёӨдёӘеӯҗ AgentгҖӮ
                        еҰӮжһңд»»еҠЎе…Ғи®ёпјҢиҜ·еҠЎеҝ…е№¶иЎҢи°ғз”Ёе®ғд»¬д»ҘжҸҗй«ҳж•ҲзҺҮгҖӮ
                        жңҖеҗҺжұҮжҖ»е®ғд»¬зҡ„жҠҘе‘ҠгҖӮ"""
    )

    task = "иҜ·иҜҰз»Ҷи°ғз ” 'LangChain DeepAgents' жЎҶжһ¶гҖӮжҲ‘йңҖиҰҒе®ҳж–№зҡ„жҠҖжңҜжһ¶жһ„иҜҙжҳҺпјҲжқҘиҮӘж–ҮжЎЈпјүд»ҘеҸҠзӨҫеҢәзҡ„жңҖдҪіе®һи·өжЎҲдҫӢгҖӮиҜ·еҜ№жҜ”дёӨиҖ…гҖӮ"

    console.print(f"\n[bold green]д»»еҠЎжҢҮд»Ө:[/bold green] {task}\n")

    # иҝҗиЎҢе№¶еҸҜи§ҶеҢ–
    step = 0
    try:
        async for event in agent.astream({"messages": [("user", task)]}):
            step += 1
            # йҒҚеҺҶжүҖжңүиҠӮзӮ№иҫ“еҮә (e.g., 'agent', 'tools')
            for node_name, node_data in event.items():
                if node_data is None:
                    continue

                if "messages" in node_data:
                    msgs = node_data["messages"]
                    # зЎ®дҝқжҳҜеҲ—иЎЁ
                    if not isinstance(msgs, list):
                        msgs = [msgs]

                    for msg in msgs:
                        # 0. иҝҮж»Өйқһж¶ҲжҒҜеҜ№иұЎ
                        if not isinstance(msg, BaseMessage):
                            continue

                        # 1. жЈҖжөӢе·Ҙе…·и°ғз”Ё (жңҹжңӣзңӢеҲ° 'task' е·Ҙе…·)
                        if hasattr(msg, "tool_calls") and msg.tool_calls:
                            tree = Tree(f"[bold yellow]Step {step}: еҶізӯ–дёҺи°ғз”Ё (Node: {node_name})[/bold yellow]")
                            for tc in msg.tool_calls:
                                tool_name = tc['name']
                                tool_args = tc['args']

                                if tool_name == "task":
                                    # йӘҢиҜҒжҲҗеҠҹпјҒ
                                    branch = tree.add(f"[bold red]рҹҡҖ и§ҰеҸ‘ 'task' е·Ҙе…· (Sub-Agent)[/bold red]")
                                    branch.add(f"[cyan]еӯҗ Agent зұ»еһӢ:[/cyan] {tool_args.get('subagent_type')}")
                                    branch.add(f"[cyan]д»»еҠЎжҢҮд»Ө:[/cyan] {tool_args.get('description')}")
                                else:
                                    tree.add(f"[blue]жҷ®йҖҡе·Ҙе…·и°ғз”Ё:[/blue] {tool_name}")

                            console.print(tree)

                        # 2. жЈҖжөӢе·Ҙе…·иҫ“еҮә (Sub-Agent зҡ„иҝ”еӣһз»“жһң)
                        elif isinstance(msg, ToolMessage):
                            if msg.name == "task":
                                # Sub-Agent е®ҢжҲҗд»»еҠЎиҝ”еӣһ
                                panel = Panel(
                                    msg.content,
                                    title=f"[bold magenta]Sub-Agent е®ҢжҲҗд»»еҠЎ (Node: {node_name})[/bold magenta]",
                                    border_style="magenta"
                                )
                                console.print(panel)
                            else:
                                console.print(f"[dim]Tool Output ({msg.name}): {msg.content[:100]}...[/dim]")

                        # 3. жЈҖжөӢ AI жңҖз»ҲеӣһеӨҚ
                        elif msg.content and not msg.tool_calls:
                            title = f"[bold green]Agent еӣһеӨҚ (Node: {node_name})[/bold green]"
                            console.print(Panel(msg.content, title=title, border_style="green"))

    except Exception as e:
        console.print(f"[bold red]вқҢ иҝҗиЎҢж—¶й”ҷиҜҜ: {e}[/bold red]")


    console.print("\n[bold blue]жј”зӨәз»“жқҹ[/bold blue]")

# иҝҗиЎҢжј”зӨә
if __name__ == "__main__":
    await run_parallel_demo()
````

# <center> з¬¬еӣӣйҳ¶ж®өгҖҒ ж–Үд»¶зі»з»ҹйӣҶжҲҗ (Filesystem & Sandbox)


- **з»„д»¶**пјҡ`FilesystemMiddleware` & `Backend`

- **еҠҹиғҪ**пјҡиҝҷжҳҜ DeepAgent зҡ„ж ёеҝғ"еӨ–жҢӮеӨ§и„‘"гҖӮAgent дёҚеҶҚе°ҶжүҖжңүжЈҖзҙўеҲ°зҡ„й•ҝж–ҮеЎһе…ҘдёҠдёӢж–ҮпјҲContextпјүпјҢиҖҢжҳҜе°Ҷе…¶еҶҷе…Ҙ FilesystemпјҲеҰӮ `/workspace`пјүпјҢд»…еңЁйңҖиҰҒж—¶иҜ»еҸ–гҖӮиҝҷжһҒеӨ§ең°жү©еұ•дәҶ Agent зҡ„"е·ҘдҪңи®°еҝҶ"гҖӮ

- **ж ёеҝғе·Ҙе…·йӣҶ**пјҡ

  - `ls`: жөҸи§Ҳзӣ®еҪ•з»“жһ„гҖӮ

  - `read_file`: иҜ»еҸ–ж–Үд»¶еҶ…е®№гҖӮ**е…ій”®е°ҒиЈ…**пјҡж”ҜжҢҒ `offset`пјҲеҒҸз§»йҮҸпјүе’Ң `limit`пјҲиЎҢж•°пјүпјҢејәеҲ¶ Agent еҜ№еӨ§ж–Үд»¶иҝӣиЎҢеҲҶйЎөиҜ»еҸ–пјҢйҒҝе…ҚдёҖж¬ЎжҖ§иҜ»зҲҶдёҠдёӢж–ҮгҖӮ

  - `write_file`: еҲӣе»әж–°ж–Үд»¶гҖӮ

  - `edit_file`: дҝ®ж”№ж–Үд»¶гҖӮж”ҜжҢҒзІҫзЎ®зҡ„еӯ—з¬ҰдёІжӣҝжҚўе’Ң `replace_all` жЁЎејҸпјҢе№¶жңүйҳІе‘Ҷи®ҫи®ЎпјҲиҰҒжұӮе…ҲиҜ»еҗҺж”№пјүгҖӮ

  - `glob`: йҖҡй…Қз¬ҰжЁЎзіҠжҹҘжүҫж–Үд»¶гҖӮ

  - `grep`: жӯЈеҲҷиЎЁиҫҫејҸжҗңзҙўж–Үд»¶еҶ…е®№пјҢеғҸе‘Ҫд»ӨиЎҢдёҖж ·еңЁд»Јз Ғеә“дёӯе®ҡдҪҚзӣ®ж ҮгҖӮ

  - `execute`: жү§иЎҢ Shell е‘Ҫд»ӨпјҲйңҖ Sandbox ж”ҜжҢҒпјүпјҢз”ЁдәҺиҝҗиЎҢд»Јз ҒгҖҒе®үиЈ…дҫқиө–зӯүгҖӮ

- **дә®зӮ№еҠҹиғҪпјҡеӨ§з»“жһңиҮӘеҠЁиҪ¬еӯҳ**

  - еҪ“ Agent и°ғз”Ёе·Ҙе…·дә§з”ҹзҡ„з»“жһңиҝҮй•ҝж—¶пјҢзі»з»ҹдјҡиҮӘеҠЁжӢҰжҲӘпјҢе°Ҷе®Ңж•ҙз»“жһңеҶҷе…Ҙж–Үд»¶зі»з»ҹпјҲеҰӮ `/large_tool_results/{id}`пјүпјҢе№¶еҸӘз»ҷ Agent иҝ”еӣһдёҖдёӘж‘ҳиҰҒе’Ңж–Үд»¶и·Ҝеҫ„гҖӮиҝҷе®ҢзҫҺи§ЈеҶідәҶжҗңзҙўз»“жһңжҲ–ж—Ҙеҝ—ж–Үд»¶иҝҮй•ҝеҜјиҮҙ Agent еҙ©жәғзҡ„й—®йўҳгҖӮ

- **еҗҺз«Ҝж”ҜжҢҒ (Backends)**пјҡ

  - **FilesystemBackend**: зӣҙжҺҘж“ҚдҪңжң¬ең°зЈҒзӣҳгҖӮ

  - **DockerBackend**: еңЁ Docker е®№еҷЁдёӯжү§иЎҢпјҢжҸҗдҫӣйҡ”зҰ»зҺҜеўғгҖӮ

  - **StoreBackend**: еӯҳеӮЁеңЁж•°жҚ®еә“дёӯпјҢж”ҜжҢҒжҹҘиҜўе’ҢжЈҖзҙўгҖӮ

  - **E2BBackend**: дҪҝз”Ё E2B дә‘з«ҜжІҷз®ұгҖӮ

  - **CompositeBackend**: ж··еҗҲжЁЎејҸпјҲеҰӮпјҡжң¬ең°еӯҳж–Үд»¶пјҢDocker и·‘д»Јз ҒпјүгҖӮ

## дёҖгҖҒ ж ёеҝғе·Ҙе…·йӣҶжөӢиҜ•

````python
# --- FilesystemMiddleware е·Ҙе…·жөӢиҜ• ---
print("\n" + "="*50)
print("рҹҡҖ ејҖе§ӢжөӢиҜ• FilesystemMiddleware е·Ҙе…·йӣҶ")
print("="*50 + "\n")

from deepagents import create_deep_agent
from deepagents.backends import FilesystemBackend
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv

# 1. еҲқе§ӢеҢ–зҺҜеўғ
load_dotenv(override=True)
model = ChatOpenAI(model="gpt-4o", temperature=0)

# дҪҝз”Ёжң¬ең°ж–Үд»¶зі»з»ҹеҗҺз«ҜпјҢж №зӣ®еҪ•и®ҫдёә ./workspace
# иҝҷж ·жҲ‘д»¬еҸҜд»ҘзңӢеҲ°зңҹе®һж–Үд»¶зҡ„еҲӣе»ә
# жіЁж„Ҹпјҡи®ҫзҪ® virtual_mode=True д»Ҙж”ҜжҢҒз»қеҜ№и·Ҝеҫ„ (еҰӮ /hello_world.py) жҳ е°„еҲ° ./workspace
backend = FilesystemBackend(root_dir="./workspace", virtual_mode=True)

# еҲӣе»ә Agent
# жіЁж„Ҹпјҡcreate_deep_agent й»ҳи®ӨдјҡиҮӘеҠЁеҢ…еҗ« FilesystemMiddleware
agent = create_deep_agent(
    model=model,
    backend=backend,
    system_prompt="дҪ жҳҜдёҖдёӘж–Үд»¶зі»з»ҹж“ҚдҪңеҠ©жүӢгҖӮиҜ·ж №жҚ®з”ЁжҲ·жҢҮд»ӨдҪҝз”Ёзӣёеә”зҡ„е·Ҙе…·гҖӮ"
)

# иҫ…еҠ©еҮҪж•°пјҡиҝҗиЎҢе№¶жү“еҚ°з»“жһң
def run_test(task_name, instruction):
    print(f"\nрҹ”№ [жөӢиҜ•: {task_name}]")
    print(f"жҢҮд»Ө: {instruction}")
    try:
        # дҪҝз”Ё invoke иҖҢдёҚжҳҜ stream д»Ҙз®ҖеҢ–иҫ“еҮә
        result = agent.invoke({"messages": [("user", instruction)]})
        last_msg = result["messages"][-1]
        print(f"рҹӨ– Agent еӣһеӨҚ: {last_msg.content}")

        # жү“еҚ°е·Ҙе…·и°ғз”ЁиҜҰжғ… (еҰӮжһңжңү)
        for msg in result["messages"]:
            if hasattr(msg, 'tool_calls') and msg.tool_calls:
                for tool in msg.tool_calls:
                    print(f"рҹӣ пёҸ  и°ғз”Ёе·Ҙе…·: {tool['name']} args={tool['args']}")
    except Exception as e:
        print(f"вқҢ еҸ‘з”ҹй”ҷиҜҜ: {e}")
````

### write_file жөӢиҜ•

````python
# 2. дҫқж¬ЎжөӢиҜ•еҗ„дёӘе·Ҙе…·

# жЎҲдҫӢ 1: write_file
# еҲӣе»әдёҖдёӘ Python и„ҡжң¬ж–Үд»¶
run_test("write_file",
         "иҜ·еңЁеҪ“еүҚзӣ®еҪ•дёӢеҲӣе»әдёҖдёӘеҗҚдёә 'hello_world.py' зҡ„ж–Үд»¶пјҢеҶ…е®№жҳҜпјҡ\nprint('Hello from DeepAgents!')")
````

### execute жөӢиҜ•

````python
# жЎҲдҫӢ 7: execute
# е°қиҜ•жү§иЎҢи„ҡжң¬
# жіЁж„Ҹпјҡexecute е·Ҙе…·йҖҡеёёйңҖиҰҒ SandboxBackend ж”ҜжҢҒгҖӮ
# еҰӮжһңеҪ“еүҚдҪҝз”Ёзҡ„ FilesystemBackend дёҚж”ҜжҢҒжү§иЎҢпјҢAgent дјҡж”¶еҲ°й”ҷиҜҜжҸҗзӨәжҲ–ж— жі•и°ғз”ЁгҖӮ
print("\nрҹ”№ [жөӢиҜ•: execute]")
print("жҢҮд»Ө: е°қиҜ•иҝҗиЎҢ hello_world.py и„ҡжң¬")
try:
    # жҲ‘д»¬е°қиҜ•ејәиЎҢиҰҒжұӮ Agent иҝҗиЎҢпјҢзңӢе®ғеҰӮдҪ•еҸҚеә”
    # еҰӮжһңжІЎжңү execute е·Ҙе…·пјҢAgent еҸҜиғҪдјҡиҜҙж— жі•жү§иЎҢ
    response = agent.invoke({"messages": [("user", "иҜ·дҪҝз”Ё execute е·Ҙе…·иҝҗиЎҢ python hello_world.py")]})
    print(f"рҹӨ– Agent еӣһеӨҚ: {response['messages'][-1].content}")
except Exception as e:
    print(f"вҡ пёҸ жөӢиҜ•иҜҙжҳҺ: execute е·Ҙе…·еҸҜиғҪдёҚеҸҜз”Ё (еҸ–еҶідәҺ Backend ж”ҜжҢҒ): {e}")

print("\n" + "="*50)
print("вң… жөӢиҜ•з»“жқҹ")
````

&emsp;&emsp;иғҪзңӢеҲ°еңЁFilesystemBackendдёӢпјҢжІЎжңүexecuteж–№жі•пјҢеӣ дёәFilesystemBackendжҳҜдёҖдёӘж–Үд»¶зі»з»ҹеҗҺз«ҜпјҢдёҚж”ҜжҢҒжү§иЎҢShellе‘Ҫд»ӨгҖӮжҺҘдёӢжқҘпјҢжҲ‘д»¬жөӢиҜ•дёҖдёӢе…¶д»–еҗҺз«ҜгҖӮ

* еҶ…еӯҳжІҷз®ұпјҢж”ҜжҢҒеңЁеҶ…еӯҳдёӯжү§иЎҢд»Јз ҒпјҢйҳІжӯўд»Јз ҒжіЁе…Ҙж”»еҮ»гҖӮд»Јз Ғжү§иЎҢз»“жһңдјҡиў«еӯҳеӮЁеңЁеҶ…еӯҳдёӯпјҢдёҚдјҡеҜ№жң¬ең°зҺҜеўғйҖ жҲҗеҪұе“ҚгҖӮжү§иЎҢз»“жқҹеҗҺпјҢеҶ…еӯҳдёӯзҡ„ж•°жҚ®дјҡиў«жё…йҷӨгҖӮ

````python
from deepagents import create_deep_agent
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI

# еҠ иҪҪзҺҜеўғеҸҳйҮҸ
load_dotenv(override=True)

# е®ҡд№ү LLM
model = ChatOpenAI(model="gpt-4o", temperature=0)

# й»ҳи®Өжғ…еҶөдёӢпјҢDeepAgent дјҡиҮӘеҠЁеҠ иҪҪ FilesystemMiddleware е№¶дҪҝз”ЁеҶ…еӯҳеҗҺз«Ҝ
agent = create_deep_agent(
    model=model,
    system_prompt="дҪ жҳҜдёҖдёӘж•°жҚ®еӨ„зҗҶеҠ©жүӢгҖӮ"
)

# Agent еҸҜд»ҘиҮӘз”ұеҲӣе»әж–Үд»¶гҖҒиҜ»еҸ–ж–Үд»¶пјҢдҪҶиҝҷдәӣж–Үд»¶еҸӘеӯҳеңЁдәҺеҶ…еӯҳдёӯ
# д»»еҠЎз»“жқҹеҗҺпјҢиҝҷдәӣж–Үд»¶дјҡиҮӘеҠЁж¶ҲеӨұ

agent.invoke({"messages": [("user", "еҲӣе»әдёҖдёӘеҗҚдёә test.txt зҡ„ж–Үд»¶е№¶еҶҷе…Ҙ 'Hello, World!'")]})
````

### 2гҖҒжҢҒд№…еҢ–жЁЎејҸпјҲж“ҚдҪңзңҹе®һж–Үд»¶пјү

* жҢҒд№…еҢ–жІҷз®ұпјҢж”ҜжҢҒе°Ҷд»Јз Ғжү§иЎҢз»“жһңжҢҒд№…еҢ–еҲ°жң¬ең°зЈҒзӣҳпјҢйҳІжӯўд»Јз Ғжү§иЎҢиҝҮзЁӢдёӯж•°жҚ®дёўеӨұгҖӮд»Јз Ғжү§иЎҢз»“жқҹеҗҺпјҢеҶ…еӯҳдёӯзҡ„ж•°жҚ®дјҡиў«жё…йҷӨпјҢдҪҶжҳҜзЈҒзӣҳдёҠзҡ„ж–Үд»¶дјҡиў«дҝқз•ҷгҖӮ

````python
from deepagents import create_deep_agent
from deepagents.backends import FilesystemBackend
from langchain_openai import ChatOpenAI

model = ChatOpenAI(model="gpt-4o", temperature=0)

# е°Ҷ Agent зҡ„ж №зӣ®еҪ•жҳ е°„еҲ°жң¬ең°зҡ„ "./workspace" ж–Үд»¶еӨ№пјҢvirtual_mode=True иЎЁзӨәеҗҜз”ЁиҷҡжӢҹж–Үд»¶зі»з»ҹпјҢдҪңз”ЁжҳҜе°Ҷ Agent жү§иЎҢзҡ„жүҖжңүж–Үд»¶ж“ҚдҪңйғҪжҳ е°„еҲ°жң¬ең°зҡ„ "./workspace" ж–Үд»¶еӨ№пјҢиҖҢдёҚжҳҜзӣҙжҺҘж“ҚдҪңжң¬ең°ж–Үд»¶зі»з»ҹгҖӮ
backend = FilesystemBackend(root_dir="./workspace", virtual_mode=True)

agent = create_deep_agent(
    model=model,
    backend=backend  # дј е…ҘеҗҺз«ҜпјҢдёӯй—ҙд»¶дјҡиҮӘеҠЁдҪҝз”Ёе®ғ
)

# жӯӨж—¶ Agent жү§иЎҢ write_file("/readme.md", ...) дјҡеңЁжң¬ең° ./workspace/readme.md еҲӣе»әж–Үд»¶
agent.invoke({"messages": [("user", "еҲӣе»әдёҖдёӘеҗҚдёә readme.md зҡ„ж–Үд»¶е№¶еҶҷе…Ҙ 'Hello, World!'")]})
````

````python
result = agent.invoke({"messages": [("user", "иҜ»еҸ– readme.md ж–Үд»¶зҡ„еҶ…е®№")]})
print(result["messages"][-1].content)
````

### 3гҖҒжј”зӨәеӨ§ж–Үд»¶иҜ»еҸ–еҲҶйЎө

* еңЁFilesystemBackendеҲҶйЎөзҡ„еҸӮж•°жҳҜеӣәе®ҡжӯ»зҡ„пјҢй»ҳи®ӨжҜҸж¬ЎиҜ»еҸ–500иЎҢпјҢжҲ‘д»¬еҸҜд»ҘеңЁзі»з»ҹжҸҗзӨәиҜҚдёӯжҢҮе®ҡжҜҸж¬ЎиҜ»еҸ–зҡ„иЎҢж•°гҖӮ

````python
import asyncio
import os
import shutil
from pathlib import Path
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.messages import BaseMessage, ToolMessage
from deepagents import create_deep_agent
from deepagents.backends import FilesystemBackend

load_dotenv(override=True)

# е®ҡд№үе·ҘдҪңзӣ®еҪ•
WORK_DIR = Path("workspace/pagination_demo").resolve()
LARGE_FILE_NAME = "server_logs.txt"
LARGE_FILE_PATH = WORK_DIR / LARGE_FILE_NAME
TARGET_SECRET = "CRITICAL_ERROR_CODE_998877"

async def run_pagination_demo():
    print("\n" + "="*80)
    print("DeepAgents FilesystemMiddleware еҲҶйЎөиҜ»еҸ–жј”зӨә")
    print("="*80)

    # 2. еҲқе§ӢеҢ– FilesystemBackend
    # е°Ҷ backend жҢҮеҗ‘жҲ‘д»¬зҡ„жөӢиҜ•зӣ®еҪ•
    # virtual_mode=True зЎ®дҝқ Agent еҸӘиғҪи®ҝй—®иҜҘзӣ®еҪ•дёӢзҡ„ж–Үд»¶пјҢдёҚиғҪи®ҝй—®е®ҝдё»жңәе…¶д»–зӣ®еҪ•
    backend = FilesystemBackend(root_dir=WORK_DIR, virtual_mode=True)

    # 3. еҲӣе»ә Agent
    # жҲ‘д»¬жҳҺзЎ®жҢҮзӨә Agent дҪҝз”ЁеҲҶйЎөиҜ»еҸ–пјҢжҜҸж¬ЎиҜ»еҸ– 300 иЎҢпјҢAgentй»ҳи®ӨжҜҸж¬ЎиҜ»еҸ–500иЎҢ(DEFAULT_READ_LIMIT)
    system_prompt = """
    дҪ жҳҜдёҖдёӘдё“дёҡзҡ„зі»з»ҹз®ЎзҗҶе‘ҳгҖӮ
    дҪ зҡ„д»»еҠЎжҳҜд»Һж—Ҙеҝ—ж–Үд»¶дёӯжҹҘжүҫзү№е®ҡзҡ„й”ҷиҜҜд»Јз ҒгҖӮ
    жіЁж„Ҹпјҡж—Ҙеҝ—ж–Үд»¶еҸҜиғҪйқһеёёеӨ§пјҢдёәдәҶйҒҝе…ҚдёҠдёӢж–ҮжәўеҮәпјҢдҪ еҝ…йЎ»дҪҝз”Ё `read_file` е·Ҙе…·зҡ„еҲҶйЎөеҠҹиғҪгҖӮ
    жҜҸж¬ЎиҜ»еҸ–иҜ·йҷҗеҲ¶еңЁ 300 иЎҢд»ҘеҶ… (limit=300)пјҢе№¶дҪҝз”Ё offset еҸӮж•°еҗ‘еҗҺж»ҡеҠЁгҖӮ
    зӣҙеҲ°жүҫеҲ°зӣ®ж ҮдҝЎжҒҜдёәжӯўгҖӮ
    """

    llm = ChatOpenAI(model="gpt-4o", temperature=0)

    agent = create_deep_agent(
        model=llm,
        backend=backend,
        system_prompt=system_prompt
    )

    task = f"иҜ·еңЁ '/{LARGE_FILE_NAME}' дёӯжҹҘжүҫеҢ…еҗ« '{TARGET_SECRET}' зҡ„иЎҢпјҢе№¶е‘ҠиҜүжҲ‘е®ғзҡ„е…·дҪ“еҶ…е®№гҖӮ"

    print(f"\nд»»еҠЎ: {task}")
    print("-" * 60)

    # 4. жү§иЎҢд»»еҠЎе№¶и§ӮеҜҹеҲҶйЎөиЎҢдёә
    step = 0
    print("\nејҖе§ӢжөҒејҸиҫ“еҮә...")
    try:
        async for event in agent.astream({"messages": [("user", task)]}):
            for node_name, node_data in event.items():
                # debug: print(f"DEBUG: Node: {node_name}")
                if not node_data: continue

                # еӨ„зҗҶ Overwrite еҜ№иұЎ
                if hasattr(node_data, "value"):
                    node_data = node_data.value

                if not isinstance(node_data, dict):
                    continue

                if "messages" in node_data:
                    msgs = node_data["messages"]
                    if hasattr(msgs, "value"):
                        msgs = msgs.value

                    if not isinstance(msgs, list): msgs = [msgs]

                    for msg in msgs:
                        # 1. жү“еҚ° Agent зҡ„жҖқиҖғ (AIMessage with tool_calls)
                        if hasattr(msg, "tool_calls") and msg.tool_calls:
                            step += 1
                            print(f"\n[Step {step}] Agent еҶіе®ҡи°ғз”Ёе·Ҙе…· (Node: {node_name}):")
                            for tc in msg.tool_calls:
                                name = tc['name']
                                args = tc['args']
                                print(f"  >>> е·Ҙе…·: {name}")

                                if name == "read_file":
                                    offset = args.get('offset', 0)
                                    limit = args.get('limit', 'Default')
                                    path_val = args.get('path') or args.get('file_path')
                                    print(f"  >>> еҸӮж•°: path='{path_val}', offset={offset}, limit={limit}")
                                    print(f"      (иҜҙжҳҺ: жӯЈеңЁиҜ»еҸ–д»Һз¬¬ {offset} иЎҢејҖе§Ӣзҡ„ {limit} иЎҢж•°жҚ®)")
                                else:
                                    print(f"  >>> еҸӮж•°: {args}")

                        # 2. жү“еҚ°е·Ҙе…·зҡ„иҫ“еҮә (ToolMessage)
                        elif isinstance(msg, ToolMessage):
                            content = msg.content
                            line_count = len(content.splitlines())
                            # жЈҖжҹҘжҳҜеҗҰеҢ…еҗ«зӣ®ж Ү Secret
                            found_secret = TARGET_SECRET in content

                            preview = content[:100].replace('\n', ' ') + "..."
                            print(f"\n[Tool Output] (Node: {node_name}) иҜ»еҸ–дәҶ {line_count} иЎҢж•°жҚ®")
                            print(f"  еҶ…е®№йў„и§Ҳ: {preview}")
                            if found_secret:
                                print(f"  вңЁ жҲҗеҠҹ: еңЁжӯӨеҲҶеқ—дёӯеҸ‘зҺ°дәҶзӣ®ж Ү Secret: {TARGET_SECRET}")

                        # 3. жү“еҚ° Agent зҡ„жңҖз»ҲеӣһеӨҚ (AIMessage without tool_calls)
                        elif isinstance(msg, BaseMessage) and msg.type == "ai" and msg.content:
                            print(f"\n[Agent жңҖз»ҲеӣһеӨҚ] (Node: {node_name}):")
                            print("-" * 40)
                            print(msg.content)
                            print("-" * 40)


    except Exception as e:
        print(f"вқҢ иҝҗиЎҢеҮәй”ҷ: {e}")
        import traceback
        traceback.print_exc()

if __name__ == "__main__":
    await run_pagination_demo()
````

````python
# е®ҡд№ү E2B жІҷз®ұеҗҺз«Ҝзұ»
# иҜҘзұ»з»§жүҝиҮӘ BaseSandboxпјҢе®һзҺ°дәҶеңЁ E2B жІҷз®ұдёӯжү§иЎҢе‘Ҫд»Өзҡ„еҠҹиғҪ

import base64
from typing import Any, Optional

# еҜје…Ҙ DeepAgents еҗҺз«ҜеҚҸи®®пјҢе®ҡд№үдәҶжІҷз®ұеҗҺз«Ҝзҡ„жҺҘеҸЈ
from deepagents.backends.protocol import (
    ExecuteResponse,
    FileDownloadResponse,
    FileUploadResponse,
    SandboxBackendProtocol,
)
# еҜје…ҘеҹәзЎҖжІҷз®ұзұ»пјҢз”ЁдәҺе®һзҺ°жІҷз®ұеҗҺз«Ҝзҡ„еҹәжң¬еҠҹиғҪ
from deepagents.backends.sandbox import BaseSandbox

try:
    from e2b import Sandbox
except ImportError:
    Sandbox = None

class E2BBackend(BaseSandbox):
    """E2B жІҷз®ұеҗҺз«Ҝе®һзҺ°пјҢз”ЁдәҺ DeepAgentsгҖӮ

    иҜҘеҗҺз«ҜдҪҝз”Ё E2BпјҲhttps://e2b.devпјүжҸҗдҫӣе®үе…ЁгҖҒйҡ”зҰ»зҡ„жү§иЎҢзҺҜеўғгҖӮ
    """

    def __init__(
        self,
        template: str = "base",
        api_key: Optional[str] = None,
        timeout: Optional[int] = None,
        metadata: Optional[dict[str, str]] = None,
    ) -> None:
        """еҲқе§ӢеҢ– E2B жІҷз®ұгҖӮ

        еҸӮж•°:
            template: E2B жІҷз®ұжЁЎжқҝ IDпјҲй»ҳи®Өпјҡ"base"пјү
            api_key: E2B API еҜҶй’ҘпјҲеҸҜйҖүпјҢй»ҳи®ӨдҪҝз”Ё E2B_API_KEY зҺҜеўғеҸҳйҮҸпјү
            timeout: жІҷз®ұи¶…ж—¶ж—¶й—ҙпјҲз§’пјү
            metadata: иҮӘе®ҡд№үжІҷз®ұе…ғж•°жҚ®
        """
        if Sandbox is None:
            raise ImportError(
                "e2b package is not installed. "
                "Please install it with `pip install e2b`."
            )

        self.sandbox = Sandbox.create(
            template=template,
            api_key=api_key,
            timeout=timeout,
            metadata=metadata,
        )

    @property
    def id(self) -> str:
        """Unique identifier for the sandbox backend."""
        return self.sandbox.sandbox_id

    def execute(self, command: str) -> ExecuteResponse:
        """Execute a command in the sandbox."""
        try:
            # E2B commands.run returns CommandResult with stdout, stderr, exit_code
            result = self.sandbox.commands.run(command)

            # иҝ”еӣһжү§иЎҢз»“жһңпјҢеҢ…еҗ« stdout жҳҜж ҮеҮҶиҫ“еҮәпјҢstderr жҳҜж ҮеҮҶй”ҷиҜҜиҫ“еҮәпјҢexit_code жҳҜйҖҖеҮәз Ғ
            return ExecuteResponse(
                output=result.stdout + result.stderr,
                exit_code=result.exit_code,
                truncated=False,
            )
        except Exception as e:
            return ExecuteResponse(
                output=f"Error executing command: {str(e)}",
                exit_code=1,
                truncated=False,
            )

    def upload_files(self, files: list[tuple[str, bytes]]) -> list[FileUploadResponse]:
        """Upload multiple files to the sandbox."""
        responses = []
        for path, content in files:
            try:
                # Ensure directory exists before writing
                # We can use execute to mkdir -p
                parent_dir = path.rsplit("/", 1)[0]
                if parent_dir:
                    self.sandbox.commands.run(f"mkdir -p {parent_dir}")

                # Write file
                self.sandbox.files.write(path, content)
                responses.append(FileUploadResponse(path=path, error=None))
            except Exception as e:
                error_msg = str(e).lower()
                error = "invalid_path"
                if "permission" in error_msg:
                    error = "permission_denied"

                responses.append(FileUploadResponse(path=path, error=error))
        return responses

    def download_files(self, paths: list[str]) -> list[FileDownloadResponse]:
        """Download multiple files from the sandbox."""
        responses = []
        for path in paths:
            try:
                content = self.sandbox.files.read(path)
                # Ensure content is bytes
                if isinstance(content, str):
                    content = content.encode("utf-8")

                responses.append(FileDownloadResponse(path=path, content=content, error=None))
            except Exception as e:
                error_msg = str(e).lower()
                error = "invalid_path"
                if "not found" in error_msg:
                    error = "file_not_found"
                elif "directory" in error_msg:
                    error = "is_directory"
                elif "permission" in error_msg:
                    error = "permission_denied"

                responses.append(FileDownloadResponse(path=path, content=None, error=error))
        return responses

    def close(self):
        """Close the sandbox session."""
        self.sandbox.kill()
````

* зҺҜеўғдёҖиҮҙжҖ§ (Reproducibility)

    - дјҳеҠҝ пјҡж— и®әжӮЁзҡ„з”өи„‘е®үиЈ…зҡ„жҳҜ Python 3.9 иҝҳжҳҜ 3.12пјҢAgent ж°ёиҝңеңЁдёҖдёӘе№ІеҮҖгҖҒж ҮеҮҶзҡ„ Python 3.11 зҺҜеўғдёӯиҝҗиЎҢгҖӮиҝҷи§ЈеҶідәҶвҖңеңЁжҲ‘зҡ„жңәеҷЁдёҠиғҪи·‘вҖқзҡ„з»Ҹе…ёдҫқиө–й—®йўҳгҖӮ

````python

#иҝҷдёӘ Backend дјҡеңЁжң¬ең°еҗҜеҠЁдёҖдёӘ Docker е®№еҷЁпјҢе№¶е°ҶдјҡиҜқйҡ”зҰ»еңЁе®№еҷЁеҶ…йғЁгҖӮ

# - ж ёеҝғеҠҹиғҪ пјҡ
#   - иҮӘеҠЁз”ҹе‘Ҫе‘Ёжңҹз®ЎзҗҶ пјҡеҲқе§ӢеҢ–ж—¶еҗҜеҠЁе®№еҷЁпјҢз»“жқҹж—¶иҮӘеҠЁй”ҖжҜҒ ( auto_remove=True )гҖӮ
#   - й«ҳж•Ҳж–Үд»¶дј иҫ“ пјҡдҪҝз”Ё tar жөҒеңЁе®ҝдё»жңәе’Ңе®№еҷЁд№Ӣй—ҙдј иҫ“ж–Үд»¶пјҢж”ҜжҢҒжү№йҮҸж“ҚдҪңгҖӮ
#   - иө„жәҗйҷҗеҲ¶ пјҡж”ҜжҢҒи®ҫзҪ® CPU ( cpu_quota ) е’ҢеҶ…еӯҳ ( memory_limit ) йҷҗеҲ¶пјҢйҳІжӯў Agent иҖ—е°Ҫжң¬жңәиө„жәҗгҖӮ
#   - зҪ‘з»ңжҺ§еҲ¶ пјҡеҸҜйҖүзҰҒз”ЁзҪ‘з»ң ( network_disabled=True ) д»Ҙеўһејәе®үе…ЁжҖ§гҖӮ

import io
import tarfile
import time
import uuid
from typing import Optional

# еҠ иҪҪ DeepAgents еҗҺз«ҜеҚҸи®®
from deepagents.backends.protocol import (
    ExecuteResponse,
    FileDownloadResponse,
    FileUploadResponse,
    SandboxBackendProtocol,
)

# еҠ иҪҪ DeepAgents еҹәзЎҖжІҷз®ұзұ»
from deepagents.backends.sandbox import BaseSandbox

try:
    import docker
    from docker.errors import NotFound, APIError
except ImportError:
    docker = None

class DockerBackend(BaseSandbox):
    """Docker жІҷз®ұеҗҺз«Ҝе®һзҺ°пјҢз”ЁдәҺ DeepAgentsгҖӮ

    иҜҘеҗҺз«ҜдҪҝз”Ёжң¬ең° Docker е®ҲжҠӨиҝӣзЁӢжҸҗдҫӣйҡ”зҰ»зҡ„жү§иЎҢзҺҜеўғгҖӮ
    йңҖиҰҒе®үиЈ… `docker` Python еҢ…пјҢе№¶зЎ®дҝқ Docker е®ҲжҠӨиҝӣзЁӢжӯЈеңЁиҝҗиЎҢгҖӮ
    """

    def __init__(
        self,
        image: str = "python:3.11-slim",
        auto_remove: bool = True,
        cpu_quota: int = 50000,  # 50% CPU
        memory_limit: str = "512m",
        network_disabled: bool = False,
        working_dir: str = "/workspace",
        volumes: dict[str, dict[str, str]] | None = None,
    ) -> None:
        """еҲқе§ӢеҢ– Docker жІҷз®ұгҖӮ

        еҸӮж•°:
            image: дҪҝз”Ёзҡ„ Docker й•ңеғҸпјҲй»ҳи®Өпјҡ"python:3.11-slim"пјү
            auto_remove: жҳҜеҗҰеңЁе…ій—ӯж—¶з§»йҷӨе®№еҷЁпјҲй»ҳи®ӨпјҡTrueпјү
            cpu_quota: CPU й…ҚйўқпјҢеҚ•дҪҚдёәеҫ®з§’пјҲй»ҳи®Өпјҡ50000пјү
            memory_limit: еҶ…еӯҳйҷҗеҲ¶пјҲй»ҳи®Өпјҡ"512m"пјү
            network_disabled: жҳҜеҗҰзҰҒз”ЁзҪ‘з»ңи®ҝй—®пјҲй»ҳи®ӨпјҡFalseпјү
            working_dir: е®№еҷЁеҶ…зҡ„е·ҘдҪңзӣ®еҪ•пјҲй»ҳи®Өпјҡ"/workspace"пјү
            volumes: Docker еҚ·й…ҚзҪ®пјҢдҫӢеҰӮ {'/е®ҝдё»жңәи·Ҝеҫ„': {'bind': '/е®№еҷЁи·Ҝеҫ„', 'mode': 'rw'}}
        """
        if docker is None:
            raise ImportError(
                "docker package is not installed. "
                "Please install it with `pip install docker`."
            )
        # еҲқе§ӢеҢ– Docker е®ўжҲ·з«ҜпјҢfrom_env() дјҡиҮӘеҠЁд»ҺзҺҜеўғеҸҳйҮҸдёӯиҜ»еҸ– Docker й…ҚзҪ®
        self.client = docker.from_env()
        self.image = image
        self.auto_remove = auto_remove
        self.working_dir = working_dir
        self.volumes = volumes or {}
        self._container = None

        # Start container
        try:
            # Ensure image exists
            try:
                self.client.images.get(image)
            except NotFound:
                print(f"Pulling image {image}...")
                self.client.images.pull(image)

            self._container = self.client.containers.run(
                image,
                command="tail -f /dev/null",  # Keep container running
                detach=True,
                tty=True,
                cpu_quota=cpu_quota,
                mem_limit=memory_limit,
                network_disabled=network_disabled,
                working_dir=working_dir,
                volumes=self.volumes,
            )

            # Ensure working directory exists
            self.execute(f"mkdir -p {working_dir}")

        except Exception as e:
            raise RuntimeError(f"Failed to start Docker container: {e}")

    @property
    def id(self) -> str:
        """Unique identifier for the sandbox backend."""
        return self._container.id if self._container else "unknown"

    def execute(self, command: str) -> ExecuteResponse:
        """Execute a command in the sandbox."""
        if not self._container:
            return ExecuteResponse(
                output="Container not running",
                exit_code=1,
                truncated=False
            )

        try:
            # Docker exec_run иҝ”еӣһ (exit_code, output)
            # output жҳҜеӯ—иҠӮзұ»еһӢ
            # дҪҝз”ЁеҲ—иЎЁеҪўејҸзҡ„ cmd д»ҘйҒҝе…Қ shell иҪ¬д№үй—®йўҳ
            exit_code, output = self._container.exec_run(
                cmd=["bash", "-c", command],
                workdir=self.working_dir,
                demux=False # Combine stdout and stderr
            )

            return ExecuteResponse(
                output=output.decode("utf-8", errors="replace"),
                exit_code=exit_code,
                truncated=False,
            )
        except Exception as e:
            return ExecuteResponse(
                output=f"Error executing command: {str(e)}",
                exit_code=1,
                truncated=False,
            )

    def upload_files(self, files: list[tuple[str, bytes]]) -> list[FileUploadResponse]:
        """Upload multiple files to the sandbox using tar archive."""
        if not self._container:
            return [FileUploadResponse(path=p, error="permission_denied") for p, _ in files]

        responses = []

        # Create a tar archive in memory
        tar_stream = io.BytesIO()
        with tarfile.open(fileobj=tar_stream, mode='w') as tar:
            for path, content in files:
                # Docker put_archive expects relative paths inside the tar to be relative to the destination
                # But here we want absolute paths to be respected.
                # Actually put_archive extracts to a directory.
                # To support absolute paths, we should probably upload to root /?
                # Or handle relative paths relative to working_dir.

                # Let's handle paths:
                # If path is absolute, we strip leading / and upload to root.
                # If path is relative, we upload to working_dir.

                # Simplification: We will create a tar with full structure and extract to /

                # Normalize path
                if path.startswith("/"):
                    arcname = path.lstrip("/")
                    dest_path = "/"
                else:
                    arcname = path
                    dest_path = self.working_dir

                info = tarfile.TarInfo(name=arcname)
                info.size = len(content)
                info.mtime = time.time()
                tar.addfile(info, io.BytesIO(content))

                responses.append(FileUploadResponse(path=path, error=None))

        tar_stream.seek(0)

        try:
            # We extract to / to support absolute paths in the tar
            # Note: This assumes all files in the batch can be extracted to the same root.
            # If mixed absolute/relative, this might be tricky.
            # For robustness, we might need to upload one by one if paths are mixed,
            # or group them.
            # Strategy: Always extract to / (root), and ensure arcnames are full paths (without leading /)

            self._container.put_archive(
                path="/",
                data=tar_stream
            )
        except Exception as e:
            # Mark all as failed if batch fails
            return [FileUploadResponse(path=p, error="permission_denied") for p, _ in files]

        return responses

    def download_files(self, paths: list[str]) -> list[FileDownloadResponse]:
        """Download multiple files from the sandbox."""
        if not self._container:
            return [FileDownloadResponse(path=p, error="permission_denied") for p in paths]

        responses = []
        for path in paths:
            try:
                # get_archive returns a tuple (generator, stat)
                bits, stat = self._container.get_archive(path)

                # Reconstruct tar from bits
                file_content = io.BytesIO()
                for chunk in bits:
                    file_content.write(chunk)
                file_content.seek(0)

                # Extract file from tar
                with tarfile.open(fileobj=file_content, mode='r') as tar:
                    # There should be only one file/dir
                    member = tar.next()
                    if member.isdir():
                        responses.append(FileDownloadResponse(path=path, error="is_directory"))
                        continue

                    f = tar.extractfile(member)
                    if f:
                        content = f.read()
                        responses.append(FileDownloadResponse(path=path, content=content, error=None))
                    else:
                        responses.append(FileDownloadResponse(path=path, error="file_not_found"))

            except NotFound:
                responses.append(FileDownloadResponse(path=path, error="file_not_found"))
            except Exception as e:
                error_msg = str(e).lower()
                error = "invalid_path"
                if "permission" in error_msg:
                    error = "permission_denied"

                responses.append(FileDownloadResponse(path=path, content=None, error=error))
        return responses

    def close(self):
        """Close the sandbox session."""
        if self._container:
            try:
                if self.auto_remove:
                    self._container.remove(force=True)
                else:
                    self._container.stop()
            except Exception:
                pass
            self._container = None
````

````python
import asyncio
import os
import sys
from dotenv import load_dotenv
from deepagents import create_deep_agent
from langchain_openai import ChatOpenAI
from langchain_core.messages import BaseMessage

# е°қиҜ•еҜје…Ҙ DockerBackend
try:
    from docker_backend import DockerBackend
except ImportError:
    sys.path.append(os.path.dirname(os.path.abspath(__file__)))
    from docker_backend import DockerBackend

load_dotenv(override=True)

async def run_docker_demo():
    print("\n" + "="*80)
    print("DeepAgents DockerBackend жј”зӨә (жһҒз®ҖзүҲ)")
    print("="*80)

    # 1. жЈҖжҹҘ Docker зҺҜеўғ
    try:
        import docker
        docker.from_env().ping()
    except Exception as e:
        print(f"вқҢ Docker жңӘиҝҗиЎҢжҲ–жңӘе®үиЈ…: {e}")
        print("иҜ·зЎ®дҝқ Docker Desktop е·ІеҗҜеҠЁгҖӮ")
        return

    # 2. еҲқе§ӢеҢ– Docker Backend
    print("жӯЈеңЁеҗҜеҠЁ Docker е®№еҷЁ (Image: python:3.11-slim)...")
    try:
        backend = DockerBackend(
            image="python:3.11-slim",
            auto_remove=True
        )
        print(f"е®№еҷЁе·ІеҗҜеҠЁ (ID: {backend.id[:12]})")
    except Exception as e:
        print(f"вқҢ еҗҜеҠЁеӨұиҙҘ: {e}")
        return

    try:
        # 3. еҲӣе»ә Agent
        llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

        agent = create_deep_agent(
            model=llm,
            backend=backend,
            system_prompt="""дҪ жҳҜдёҖдёӘиҝҗиЎҢеңЁ Docker е®№еҷЁдёӯзҡ„ AI еҠ©жүӢгҖӮ
            дҪ зҡ„д»»еҠЎжҳҜжј”зӨәзҺҜеўғйҡ”зҰ»жҖ§гҖӮ

            иҜ·жү§иЎҢд»ҘдёӢжӯҘйӘӨпјҡ
            1. иҝҗиЎҢ 'cat /etc/os-release' жҹҘзңӢе®№еҷЁж“ҚдҪңзі»з»ҹгҖӮ
            2. иҝҗиЎҢ 'python --version' зЎ®и®Ө Python зҺҜеўғгҖӮ
            3. еҲӣе»әж–Үд»¶ '/workspace/hello_docker.py'пјҢеҶ…е®№дёәжү“еҚ° 'Hello from Docker Container!'гҖӮ
            4. иҝҗиЎҢиҜҘи„ҡжң¬гҖӮ
            """
        )

        print(f"\nд»»еҠЎејҖе§Ӣ")

        # 4. жү§иЎҢд»»еҠЎ
        async for event in agent.astream({"messages": [("user", "иҜ·ејҖе§Ӣжј”зӨәгҖӮ")]}):
            for node_name, node_data in event.items():
                if not node_data:
                    continue

                # еӨ„зҗҶ Overwrite еҜ№иұЎ
                if hasattr(node_data, "value"):
                    node_data = node_data.value

                # еҶҚж¬ЎжЈҖжҹҘжҳҜеҗҰдёәеӯ—е…ё
                if not isinstance(node_data, dict):
                    continue

                if "messages" in node_data:
                    messages = node_data["messages"]
                    if hasattr(messages, "value"):
                        messages = messages.value

                    if isinstance(messages, list) and messages:
                        last_msg = messages[-1]
                        if isinstance(last_msg, BaseMessage) and last_msg.content:
                            print(f"\n[Agent ({node_name})]")
                            print("-" * 40)
                            print(last_msg.content)
                            print("-" * 40)

    except Exception as e:
        print(f"вқҢ иҝҗиЎҢж—¶й”ҷиҜҜ: {e}")
        import traceback
        traceback.print_exc()
    finally:
        print("\nжӯЈеңЁжё…зҗҶе®№еҷЁ...")
        backend.close()
        print("е®№еҷЁе·Із§»йҷӨ")
        print("\nжј”зӨәз»“жқҹ")

if __name__ == "__main__":
    await run_docker_demo()
````

````python
import asyncio
import os
import uuid
import traceback
from dotenv import load_dotenv

# LangChain / LangGraph Imports
from langchain_openai import ChatOpenAI
from langchain_mcp_adapters.client import MultiServerMCPClient
from langchain_core.messages import BaseMessage, ToolMessage
from langgraph.store.postgres import PostgresStore
from langgraph.checkpoint.memory import MemorySaver
from psycopg_pool import ConnectionPool

# DeepAgents Imports
from deepagents import create_deep_agent
from deepagents.backends import StoreBackend

# еҠ иҪҪзҺҜеўғеҸҳйҮҸ
load_dotenv(override=True)

# ж•°жҚ®еә“иҝһжҺҘеӯ—з¬ҰдёІ (иҜ·ж №жҚ®е®һйҷ…жғ…еҶөдҝ®ж”№)
DB_URI = "postgresql://myuser:123456@localhost:5432/mydatabase"

def print_header():
    print("\n" + "="*80)
    print("DeepAgents StoreBackend (PostgreSQL) жј”зӨә (жһҒз®ҖзүҲ)")
    print("="*80)

async def setup_mcp_tools():
    """
    иҝһжҺҘ Context7 MCP жңҚеҠЎеҷЁе№¶иҺ·еҸ–е·Ҙе…·гҖӮ
    """
    print("Step 1.1: жӯЈеңЁиҝһжҺҘ Context7 MCP жңҚеҠЎеҷЁ...")
    try:
        # дҪҝз”Ёе®ҳж–№ Context7 MCP й…ҚзҪ®
        client = MultiServerMCPClient({
            "context7": {
                "transport": "stdio",
                "command": "npx",
                "args": ["-y", "@upstash/context7-mcp@latest"],
            }
        })

        # иҺ·еҸ–е·Ҙе…·еҲ—иЎЁ
        tools = await client.get_tools()

        print(f"Step 1.2: жҲҗеҠҹеҠ иҪҪ {len(tools)} дёӘ MCP е·Ҙе…·")
        return client, tools
    except Exception as e:
        print(f"ERROR: иҝһжҺҘ MCP еӨұиҙҘ: {e}")
        return None, []


async def run_store_backend_demo():
    """
    иҝҗиЎҢ StoreBackend жј”зӨәпјҡ
    еұ•зӨәеҰӮдҪ•дҪҝз”Ё PostgreSQL дҪңдёә DeepAgents зҡ„ж–Үд»¶зі»з»ҹеҗҺз«ҜгҖӮ
    """
    print_header()

    # =================================================================================================
    # Step 1. еҲқе§ӢеҢ– MCP е·Ҙе…· (Connect to Context7 MCP Server)
    # =================================================================================================
    mcp_client, mcp_tools = await setup_mcp_tools()


    # =================================================================================================
    # Step 2. еҲқе§ӢеҢ– PostgreSQL иҝһжҺҘжұ е’ҢеӯҳеӮЁз»„д»¶ (Init DB & Store)
    # =================================================================================================
    print(f"\nStep 2: иҝһжҺҘж•°жҚ®еә“ {DB_URI} е№¶еҲқе§ӢеҢ–еӯҳеӮЁ...")

    # дҪҝз”Ё ConnectionPool з®ЎзҗҶж•°жҚ®еә“иҝһжҺҘ
    # StoreBackend йңҖиҰҒеҗҢжӯҘзҡ„ ConnectionPool (еӣ дёәе®ғеңЁзәҝзЁӢдёӯиҝҗиЎҢеҗҢжӯҘж“ҚдҪң)
    try:
        with ConnectionPool(conninfo=DB_URI, kwargs={"autocommit": True}) as pool:

            # 2.1 еҲқе§ӢеҢ– Checkpointer (з”ЁдәҺдҝқеӯҳдјҡиҜқзҠ¶жҖҒ/иҒҠеӨ©и®°еҪ•)
            # жіЁж„Ҹ: иҝҷйҮҢдҪҝз”Ё MemorySaver жҳҜдёәдәҶйҒҝејҖ langgraph-checkpoint-postgres еңЁжҹҗдәӣзҺҜеўғдёӢзҡ„ејӮжӯҘе…је®№жҖ§й—®йўҳ
            checkpointer = MemorySaver()

            # 2.2 еҲқе§ӢеҢ– Store (з”ЁдәҺдҝқеӯҳж–Үд»¶/й•ҝжңҹи®°еҝҶ)
            # DeepAgents зҡ„ StoreBackend дјҡе°Ҷж–Үд»¶зі»з»ҹж“ҚдҪңжҳ е°„еҲ°иҝҷдёӘ PostgresStore
            store = PostgresStore(pool)

            # 2.3 жү§иЎҢж•°жҚ®еә“иҝҒз§» (йҰ–ж¬ЎиҝҗиЎҢйңҖиҰҒеҲқе§ӢеҢ–иЎЁз»“жһ„)
            # иҝҷдјҡеңЁж•°жҚ®еә“дёӯеҲӣе»ә 'store' иЎЁпјҢз”ЁдәҺеӯҳеӮЁ JSON ж•°жҚ® (еҚіжҲ‘д»¬зҡ„ж–Үд»¶)
            with pool.connection() as conn:
                with conn.cursor() as cur:
                    for migration in store.MIGRATIONS:
                        cur.execute(migration)

            print("Step 2: ж•°жҚ®еә“иЎЁз»“жһ„е°ұз»Ә")


            # =================================================================================================
            # Step 3. еҲӣе»әеҹәдәҺ StoreBackend зҡ„ Agent (Create Agent)
            # =================================================================================================

            # StoreBackend е°Ҷж–Үд»¶зі»з»ҹж“ҚдҪңжҳ е°„еҲ° LangGraph зҡ„ Store
            # иҝҷж„Ҹе‘ізқҖж–Үд»¶е°ҶжҢҒд№…еҢ–еӯҳеӮЁеңЁ PostgreSQL ж•°жҚ®еә“дёӯпјҢи·ЁйҮҚеҗҜеҸҜз”Ё
            llm = ChatOpenAI(model="gpt-4o", temperature=0)

            # е®ҡд№ү Backend Factory
            # иҝҷжҳҜдёҖдёӘе·ҘеҺӮеҮҪж•°пјҢз”ЁдәҺеңЁиҝҗиЎҢж—¶(runtime)жҠҠ StoreBackend е®һдҫӢеҢ–гҖӮ
            # create_deep_agent дјҡеңЁеҶ…йғЁжҠҠ LangGraph зҡ„ runtime еҜ№иұЎдҪңдёәеҸӮж•° rt дј иҝӣжқҘпјҢ
            # иҝҷж · StoreBackend е°ұиғҪжӢҝеҲ° store/checkpointer зӯүиө„жәҗпјҢе®һзҺ°ж–Үд»¶зі»з»ҹеҲ° PostgreSQL зҡ„жҳ е°„гҖӮ
            backend_factory = lambda rt: StoreBackend(rt)
            print("Step 3: жӯЈеңЁеҲқе§ӢеҢ– DeepAgent (Backend: StoreBackend -> PostgresStore)...")

            agent = create_deep_agent(
                model=llm,
                tools=mcp_tools,
                backend=backend_factory,
                store=store,              # е…ій”®: еҝ…йЎ»дј е…Ҙ store е®һдҫӢ
                checkpointer=checkpointer, # е…ій”®: дј е…Ҙ checkpointer е®һдҫӢ
                system_prompt="""дҪ жҳҜдёҖдёӘй«ҳзә§жҠҖжңҜеҠ©жүӢгҖӮ
                дҪ зҡ„д»»еҠЎжҳҜдҪҝз”Ё Context7 е·Ҙе…·жҹҘиҜўе…ідәҺ 'DeepAgents StoreBackend' зҡ„з”Ёжі•гҖӮ
                жҹҘиҜўеҗҺпјҢеҲӣе»әдёҖдёӘжҖ»з»“ж–Үд»¶ '/knowledge/store_backend_notes.md'пјҢе№¶еҶҷе…Ҙе…ій”®дҝЎжҒҜгҖӮ
                з”ұдәҺдҪ дҪҝз”Ёзҡ„жҳҜ StoreBackendпјҢиҝҷдёӘж–Үд»¶е°ҶзӣҙжҺҘеӯҳеӮЁеңЁ PostgreSQL ж•°жҚ®еә“дёӯгҖӮ
                жңҖеҗҺпјҢиҜ·иҜ»еҸ–иҜҘж–Үд»¶д»ҘйӘҢиҜҒеӯҳеӮЁжҲҗеҠҹгҖӮ"""
            )


            # =================================================================================================
            # Step 4. жү§иЎҢд»»еҠЎ (Execute Task)
            # =================================================================================================
            thread_id = str(uuid.uuid4())
            config = {"configurable": {"thread_id": thread_id}}

            task = "иҜ·жҹҘиҜў StoreBackend зҡ„з”Ёжі•пјҢе№¶е°ҶжҖ»з»“еҶҷе…Ҙ /knowledge/store_backend_notes.mdпјҢжңҖеҗҺиҜ»еҸ–е®ғйӘҢиҜҒгҖӮ"

            # --- UI Display ------------------------------------------------------------------------------
            print(f"Session Thread ID: {thread_id}")
            print("\n" + "-"*40)
            print(f"Step 4: ејҖе§Ӣжү§иЎҢд»»еҠЎ: {task}")
            print("-"*40)
            # ---------------------------------------------------------------------------------------------

            step = 0
            message_history_len = 0

            try:
                # дҪҝз”Ё astream и§ӮеҜҹжү§иЎҢиҝҮзЁӢ
                async for event in agent.astream({"messages": [("user", task)]}, config=config):
                    if "messages" in event:
                        current_messages = event["messages"]
                        if len(current_messages) > message_history_len:
                            for i in range(message_history_len, len(current_messages)):
                                msg = current_messages[i]

                                # 1. жҳҫзӨәжҷ®йҖҡж–Үжң¬ж¶ҲжҒҜ (Agent еӣһеӨҚ)
                                if isinstance(msg, BaseMessage) and msg.content:
                                    is_tool_call = getattr(msg, "tool_calls", None)
                                    if not is_tool_call:
                                        print(f"\n[Agent]:\n{msg.content}")

                                # 2. жҳҫзӨәе·Ҙе…·и°ғз”Ё (Agent иҜ·жұӮе·Ҙе…·)
                                if hasattr(msg, "tool_calls") and msg.tool_calls:
                                    step += 1
                                    print(f"\n[Step {step}: е·Ҙе…·и°ғз”Ё]:")
                                    for tc in msg.tool_calls:
                                        print(f"  вҖў е·Ҙе…·: {tc['name']}")
                                        print(f"  вҖў еҸӮж•°: {tc['args']}")

                                # 3. жҳҫзӨәе·Ҙе…·иҫ“еҮә (Tool жү§иЎҢз»“жһң)
                                if isinstance(msg, ToolMessage):
                                    content_preview = msg.content[:200] + "..." if len(msg.content) > 200 else msg.content
                                    print(f"\n[Tool Output ({msg.name})]: {content_preview}")

                            message_history_len = len(current_messages)

            except Exception as e:
                print(f"вқҢ иҝҗиЎҢж—¶й”ҷиҜҜ: {e}")
                traceback.print_exc()


            # =================================================================================================
            # Step 5. йӘҢиҜҒжҢҒд№…еҢ–еӯҳеӮЁ (Verify Persistence)
            # =================================================================================================
            print("\n" + "="*50)
            print("Step 5: йӘҢиҜҒ StoreBackend (Postgres) жҢҒд№…еҢ–")
            print("="*50)

            print("\nйӘҢиҜҒж“ҚдҪң: дҪҝз”Ёж–° Agent е®һдҫӢ (жЁЎжӢҹйҮҚеҗҜ) иҜ»еҸ–еҗҢдёҖж–Үд»¶...")

            # жЁЎжӢҹйҮҚеҗҜпјҡеҲӣе»әдёҖдёӘж–°зҡ„ Agent е®һдҫӢпјҢдҪҶиҝһжҺҘеҗҢдёҖдёӘ Database Store
            # еҸӘиҰҒ DB иҝһжҺҘдёҚеҸҳпјҢж–Үд»¶зі»з»ҹзҠ¶жҖҒе°ұжҳҜжҢҒд№…зҡ„
            verify_agent = create_deep_agent(
                model=llm,
                backend=backend_factory,
                store=store,
                checkpointer=checkpointer,
                system_prompt="йӘҢиҜҒеҠ©жүӢ"
            )

            # иҜ»еҸ–д№ӢеүҚеҲӣе»әзҡ„ж–Үд»¶
            verify_result = await verify_agent.ainvoke({
                "messages": [("user", "иҜ·иҜ»еҸ– /knowledge/store_backend_notes.md зҡ„еҶ…е®№")]
            }, config=config)

            last_msg = verify_result["messages"][-1]

            # --- UI Display ------------------------------------------------------------------------------
            print("\n[ж•°жҚ®еә“иҜ»еҸ–йӘҢиҜҒз»“жһң]:")
            print("-" * 20)
            print(last_msg.content)
            print("-" * 20)
            print("\nиҜҙжҳҺ: 'prefix' еҜ№еә” StoreBackend зҡ„ namespace (й»ҳи®Өдёә 'filesystem')пјҢ'key' еҜ№еә”ж–Үд»¶з»қеҜ№и·Ҝеҫ„гҖӮ")

            # SQL жҹҘиҜўжҸҗзӨә
            print("\nжҸҗзӨә: дҪ еҸҜд»ҘдҪҝз”Ёд»ҘдёӢ SQL еңЁж•°жҚ®еә“дёӯзӣҙжҺҘжҹҘиҜўжӯӨж–Үд»¶:")
            sql_query = """
                        SELECT
                            key,
                            value->>'content' as content,
                            updated_at
                        FROM store
                        WHERE prefix = 'filesystem'
                        AND key = '/knowledge/store_backend_notes.md';
                                """
            print(sql_query.strip())
            # ---------------------------------------------------------------------------------------------

            print("\nжј”зӨәз»“жқҹ")

    except Exception as e:
        print(f"вқҢ ж•°жҚ®еә“иҝһжҺҘжҲ–еҲқе§ӢеҢ–еӨұиҙҘ: {e}")
        print("иҜ·жЈҖжҹҘ PostgreSQL жҳҜеҗҰжӯЈеңЁиҝҗиЎҢпјҢд»ҘеҸҠ DB_URI жҳҜеҗҰжӯЈзЎ®гҖӮ")


if __name__ == "__main__":
    # зЎ®дҝқ asyncio.run е…је®№ Jupyter зҺҜеўғ
    try:
        # asyncio.run(run_store_backend_demo())
        await run_store_backend_demo()
    except RuntimeError as e:
        if "asyncio.run() cannot be called from a running event loop" in str(e):
            print("иҜ·еңЁ Jupyter дёӯдҪҝз”Ё await run_store_backend_demo()")
        else:
            raise e
````

| еҸӮж•°з»„д»¶ | **Backend** (еҗҺз«Ҝ) | **Checkpointer** (жЈҖжҹҘзӮ№) | **Store** (еӯҳеӮЁ) |
| :--- | :--- | :--- | :--- |
| **ж ёеҝғе®ҡд№ү** | **зҺҜеўғеұӮ** (Environment) | **зҠ¶жҖҒеұӮ** (State / Short-term Memory) | **и®°еҝҶеұӮ** (Memory / Long-term Memory) |
| **иҙҹиҙЈд»Җд№Ҳпјҹ** | **"еӨ–йғЁдё–з•Ң"** зҡ„дәӨдә’иғҪеҠӣгҖӮ<br>еҚіпјҡж–Үд»¶еӯҳеңЁе“Әпјҹд»Јз ҒеңЁе“Әи·‘пјҹ | **"еҪ“еүҚеҜ№иҜқ"** зҡ„дёҠдёӢж–ҮгҖӮ<br>еҚіпјҡеҲҡжүҚиҜҙдәҶд»Җд№ҲпјҹзҺ°еңЁиҝҗиЎҢеҲ°е“ӘдёҖжӯҘдәҶпјҹ | **"и·ЁдјҡиҜқ"** зҡ„зҹҘиҜҶз§ҜзҙҜгҖӮ<br>еҚіпјҡз”ЁжҲ·еҸ«д»Җд№ҲеҗҚеӯ—пјҹдёҠж¬Ўд»»еҠЎеӯҰеҲ°дәҶд»Җд№Ҳпјҹ |
| **ж•°жҚ®зұ»еһӢ** | йқһз»“жһ„еҢ–ж–Үд»¶ (`.py`, `.md`, `.txt`)<br>иҝҗиЎҢж—¶зҺҜеўғ (Shell, Process) | `BaseMessage` еҲ—иЎЁ (User/AI/Tool Message)<br>Graph иҠӮзӮ№зҠ¶жҖҒ | з»“жһ„еҢ– JSON ж•°жҚ® (Key-Value)<br>з”ЁжҲ·еҒҸеҘҪгҖҒй•ҝжңҹз¬”и®° |
| **е…ёеһӢе®һзҺ°** | `DockerBackend` (е®№еҷЁ)<br>`FilesystemBackend` (зЈҒзӣҳ)<br>`E2BBackend` (дә‘жІҷз®ұ) | `MemorySaver` (еҶ…еӯҳ)<br>`PostgresSaver` (ж•°жҚ®еә“)<br>`SqliteSaver` (жң¬ең°DB) | `InMemoryStore` (еҶ…еӯҳ)<br>`PostgresStore` (ж•°жҚ®еә“) |
| **з”ҹе‘Ҫе‘Ёжңҹ** | **д»»еҠЎзә§**<br>(д»»еҠЎз»“жқҹе®№еҷЁеҸҜиғҪй”ҖжҜҒ) | **зәҝзЁӢзә§ (Thread)**<br>(жҚўдёӘ thread_id е°ұжІЎдәҶ) | **е…ЁеұҖзә§ (Global)**<br>(жүҖжңү thread йғҪиғҪжҹҘеҲ°) |
| **еҪўиұЎжҜ”е–»** | **е·ҘдҪңеҸ° / з”өи„‘** | **еӨ§и„‘зҡ„е·ҘдҪңи®°еҝҶ** (еҸӘдјҡжӯ»и®°зЎ¬иғҢеҪ“еүҚеҜ№иҜқ) | **ж—Ҙи®°жң¬ / зҹҘиҜҶеә“** (и®°еҪ•ж°ёд№…дҝЎжҒҜ) |

### 7гҖҒCompositeBackend дҪҝз”Ёж··еҗҲжЁЎејҸ


CompositeBackendжҳҜ Agent ж–Үд»¶ж“ҚдҪңзҡ„вҖңжҷәиғҪи·Ҝз”ұеҷЁвҖқ гҖӮ

еңЁеҚ•дёҖеҗҺз«ҜжЁЎејҸдёӢпјҢAgent жүҖжңүзҡ„ж–Үд»¶ж“ҚдҪңпјҲиҜ»гҖҒеҶҷгҖҒеҲ—еҮәзӣ®еҪ•пјүйғҪеҸӘиғҪеҺ»еҫҖеҗҢдёҖдёӘең°ж–№пјҲиҰҒд№Ҳе…ЁжҳҜжң¬ең°зЈҒзӣҳпјҢиҰҒд№Ҳе…ЁжҳҜ Docker е®№еҷЁеҶ…пјүгҖӮиҖҢ CompositeBackend е…Ғи®ёдҪ ж №жҚ® ж–Үд»¶и·Ҝеҫ„еүҚзјҖ пјҢе°ҶиҜ·жұӮеҲҶеҸ‘з»ҷдёҚеҗҢзҡ„еҗҺз«ҜгҖӮ

* ж ёеҝғдјҳеҠҝ

1. жҖ§иғҪдёҺејҖй”ҖдјҳеҢ– (Performance) иҝҷжҳҜжңҖе…ій”®зҡ„жҠҖжңҜдјҳеҠҝгҖӮ

    - DockerBackend зҡ„еұҖйҷҗ : еҗ‘ Docker е®№еҷЁеҶ…иҜ»еҶҷж–Үд»¶пјҲе°Өе…¶жҳҜеӨ§ж–Үд»¶пјүйңҖиҰҒз»ҸиҝҮ Docker Daemon зҡ„ API (еҰӮ put_archive / get_archive )пјҢж¶үеҸҠзҪ‘з»ңйҖҡдҝЎе’Ңжү“еҢ…и§ЈеҢ…пјҢејҖй”ҖиҫғеӨ§гҖӮ

    - CompositeBackend зҡ„и§Јжі• : еҜ№дәҺж•°жҚ®ж–Үд»¶пјҲ /data пјүпјҢзӣҙжҺҘйҖҡиҝҮ FilesystemBackend иҝӣиЎҢжң¬ең° I/O ж“ҚдҪңпјҢ е®Ңе…Ёз»•иҝҮдәҶ Docker API пјҢиҜ»еҶҷйҖҹеәҰжҳҜж“ҚдҪңзі»з»ҹеҺҹз”ҹзҡ„йҖҹеәҰгҖӮ

2. и®Ўз®—дёҺеӯҳеӮЁеҲҶзҰ» (Decoupling)

    - и®Ўз®—жҳҜдёҙж—¶зҡ„ : дҪ зҡ„ processor.py и„ҡжң¬еҸҜиғҪеҸӘйңҖиҰҒиҝҗиЎҢдёҖж¬ЎпјҢиҝҗиЎҢзҺҜеўғпјҲPython дҫқиө–пјүеҸҜиғҪеҫҲеӨҚжқӮдё”е®№жҳ“еҶІзӘҒгҖӮж”ҫеңЁ Docker йҮҢжңҖеҗҲйҖӮгҖӮ

    - ж•°жҚ®жҳҜж°ёжҒ’зҡ„ : дҪ зҡ„ raw_metrics.txt е’Ң health_report.txt жҳҜдёҡеҠЎиө„дә§гҖӮйҖҡиҝҮи·Ҝз”ұзӣҙжҺҘиҗҪзӣҳеҲ°е®ҝдё»жңәпјҢеҚідҪҝ Docker е®№еҷЁеҙ©жәғгҖҒиў«еҲ йҷӨжҲ–йҮҚеҗҜпјҢ ж•°жҚ®жҜ«еҸ‘ж— жҚҹ дё”з«ӢеҚіеҸҜеңЁе®ҝдё»жңәи®ҝй—®пјҲеҰӮд»Јз Ғ Line 212-222 жүҖзӨәзҡ„йӘҢиҜҒжӯҘйӘӨпјүгҖӮ
    
3. з»ҷдәҲ Agent "ж··еҗҲдә‘" зҡ„иғҪеҠӣпјҢAgent еҸҜд»ҘеғҸдәәзұ»е·ҘзЁӢеёҲдёҖж ·е·ҘдҪңпјҡ

    - "жҲ‘еңЁдёҙж—¶зҡ„жІҷз®ұйҮҢеҶҷд»Јз ҒжөӢиҜ•пјҲDockerпјүгҖӮ"

    - "жөӢиҜ•еҘҪдәҶпјҢжҲ‘жҠҠз»“жһңдҝқеӯҳеҲ°е…¬еҸёзҡ„е…ұдә«зҪ‘зӣҳйҮҢпјҲFilesystem/MountпјүгҖӮ"

````python
import asyncio
import shutil
import os
import time
from pathlib import Path
from dotenv import load_dotenv

# DeepAgents еҜје…Ҙ
from deepagents import create_deep_agent
from deepagents.backends.composite import CompositeBackend
from deepagents.backends.filesystem import FilesystemBackend
from langchain_openai import ChatOpenAI
from langchain_mcp_adapters.client import MultiServerMCPClient
from langchain_core.messages import BaseMessage, ToolMessage

# еҜје…Ҙ DockerBackend
try:
    from docker_backend import DockerBackend
except ImportError:
    try:
        from deepagents.backends.docker import DockerBackend
    except ImportError:
        DockerBackend = None

def print_header():
    print("\n" + "="*80)
    print("DeepAgents CompositeBackend ж··еҗҲеҗҺз«Ҝжј”зӨә (жһҒз®ҖзүҲ)")
    print("жһ¶жһ„пјҡж··еҗҲдә‘еҺҹз”ҹжЁЎејҸ (Docker жү§иЎҢ + жң¬ең°жҢҒд№…еҢ–)")
    print("="*80)

async def setup_mcp_tools():
    print("  вҶ’ жӯЈеңЁиҝһжҺҘ Context7 MCP жңҚеҠЎеҷЁ...")
    try:
        client = MultiServerMCPClient({
            "context7": {
                "transport": "stdio",
                "command": "npx",
                "args": ["-y", "@upstash/context7-mcp@latest"],
            }
        })
        tools = await client.get_tools()
        print("  вҶ’ MCP е·Ҙе…·еҠ иҪҪжҲҗеҠҹ")
        return client, tools
    except Exception as e:
        print(f"ERROR: MCP иҝһжҺҘеӨұиҙҘ: {e}")
        return None, []

async def run_composite_demo():
    load_dotenv(override=True)
    print_header()

    if DockerBackend is None:
        print("дёҘйҮҚй”ҷиҜҜ: жңӘжүҫеҲ° DockerBackendгҖӮиҜ·зЎ®дҝқ docker_backend.py еӯҳеңЁгҖӮ")
        return

    # Step 1
    print("\n" + "-"*40)
    print("жӯҘйӘӨ 1: й…ҚзҪ®ж··еҗҲзҺҜеўғ")
    print("-"*40)

    host_work_dir = Path("workspace/data_analysis_project").resolve()
    if host_work_dir.exists():
        shutil.rmtree(host_work_dir)
    host_work_dir.mkdir(parents=True, exist_ok=True)
    print(f"  вҖў е®ҝдё»жңәжҢҒд№…еұӮ: {host_work_dir}")

    container_mount_path = "/data"
    docker_volumes = {
        str(host_work_dir): {'bind': container_mount_path, 'mode': 'rw'}
    }
    print(f"  вҖў е®№еҷЁжҢӮиҪҪ:      {host_work_dir} вҶ” {container_mount_path}")

    # Step 2
    print("\n" + "-"*40)
    print("жӯҘйӘӨ 2: еҲқе§ӢеҢ–ж··еҗҲеҗҺз«Ҝ (Composite Backend)")
    print("-"*40)

    fs_backend = FilesystemBackend(root_dir=host_work_dir, virtual_mode=True)

    print("  вҖў жӯЈеңЁеҗҜеҠЁ Docker е®№еҷЁ (python:3.11-slim)...")
    docker_backend = DockerBackend(
        image="python:3.11-slim",
        auto_remove=True,
        volumes=docker_volumes
    )

    routes = {
        container_mount_path: fs_backend
    }
    backend = CompositeBackend(default=docker_backend, routes=routes)

    print("\n[и·Ҝз”ұиЎЁй…ҚзҪ®]")
    print(f"1. й»ҳи®Өи·Ҝз”ұ (/): DockerBackend (дёҙж—¶жү§иЎҢ)")
    print(f"2. жҢҒд№…еҢ–и·Ҝз”ұ ({container_mount_path}/*): FilesystemBackend (е®ҝдё»жңәеӯҳеӮЁ)")

    # Step 3
    print("\n" + "-"*40)
    print("жӯҘйӘӨ 3: йғЁзҪІ Agent")
    print("-"*40)

    mcp_client, mcp_tools = await setup_mcp_tools()

    system_prompt = f"""дҪ жҳҜдёҖеҗҚеңЁж··еҗҲзҺҜеўғдёӯе·ҘдҪңзҡ„й«ҳзә§ж•°жҚ®е·ҘзЁӢеёҲгҖӮ

    зҺҜеўғең°еӣҫ:
    1. жү§иЎҢеұӮ (ж №зӣ®еҪ• `/`):
       - дёҙж—¶зҡ„ Docker е®№еҷЁгҖӮ
       - з”ЁдәҺеҲӣе»әи„ҡжң¬ (`.py`) е’ҢиҝҗиЎҢе‘Ҫд»ӨгҖӮ
       - иҝҷйҮҢзҡ„ж–ҮдјҡеңЁдјҡиҜқз»“жқҹеҗҺж¶ҲеӨұгҖӮ

    2. еӯҳеӮЁеұӮ (`{container_mount_path}`):
       - д»Һе®ҝдё»жңәжҢӮиҪҪзҡ„жҢҒд№…еҢ–еӯҳеӮЁгҖӮ
       - з”ЁдәҺеӯҳж”ҫ иҫ“е…Ҙ ж•°жҚ®е’Ң иҫ“еҮә жҠҘе‘ҠгҖӮ
       - иҝҷйҮҢзҡ„ж–Үд»¶дјҡж°ёд№…дҝқеӯҳгҖӮ

    дҪ зҡ„д»»еҠЎ:
    1. **ж‘„е…Ҙ**: еҲӣе»әдёҖдёӘж–Үд»¶ `{container_mount_path}/raw_metrics.txt`пјҢеҶ…е®№дёә "CPU: 45%, Mem: 60%"гҖӮ
       (жіЁж„Ҹ: иҝҷдҪҝз”ЁдәҶ 'write_file' е·Ҙе…·пјҢиҜҘе·Ҙе…·йҖҡиҝҮи·Ҝз”ұзӣҙжҺҘеҶҷе…Ҙе®ҝдё»жңәж–Үд»¶зі»з»ҹ)гҖӮ

    2. **еӨ„зҗҶ**: еҲӣе»әдёҖдёӘ Python и„ҡжң¬ `/processor.py` (еңЁж №зӣ®еҪ•)пјҢиҜҘи„ҡжң¬:
       - иҜ»еҸ– `{container_mount_path}/raw_metrics.txt`гҖӮ
       - и®Ўз®— "еҒҘеә·еҲҶж•°" (жЁЎжӢҹдёҖдёӢеҚіеҸҜ)гҖӮ
       - е°ҶжҠҘе‘ҠеҶҷе…Ҙ `{container_mount_path}/health_report.txt`гҖӮ
       - жү“еҚ° "Analysis Complete"гҖӮ

    3. **жү§иЎҢ**: дҪҝз”Ё `python /processor.py` иҝҗиЎҢи„ҡжң¬гҖӮ
       (жіЁж„Ҹ: иҝҷеңЁ Docker еҶ…йғЁиҝҗиЎҢгҖӮDocker еӣ дёәеҚ·жҢӮиҪҪиғҪзңӢеҲ°иҝҷдәӣж–Үд»¶)гҖӮ

    4. **йӘҢиҜҒ**: иҜ»еҸ– `{container_mount_path}/health_report.txt` е№¶жҳҫзӨәе®ғгҖӮ
    """

    agent = create_deep_agent(
        model=ChatOpenAI(model="gpt-4o", temperature=0),
        tools=mcp_tools,
        backend=backend,
        system_prompt=system_prompt
    )

    # Step 4
    print("\n" + "-"*40)
    print("жӯҘйӘӨ 4: д»»еҠЎжү§иЎҢ")
    print("-"*40)

    task_input = "ејҖе§Ӣе·ҘзЁӢжөҒж°ҙзәҝгҖӮ"
    config = {"configurable": {"thread_id": "composite_demo_simple_v1"}}

    step_count = 0
    try:
        message_history_len = 0

        async for event in agent.astream({"messages": [("user", task_input)]}, config=config):
            if "messages" in event:
                current_messages = event["messages"]
                if len(current_messages) > message_history_len:
                    for i in range(message_history_len, len(current_messages)):
                        msg = current_messages[i]

                        # Agent Thinking
                        if isinstance(msg, BaseMessage) and msg.content and not getattr(msg, "tool_calls", None):
                            step_count += 1
                            print(f"\n[рҹ§  Agent жҖқиҖғ (жӯҘйӘӨ {step_count})]:\n{msg.content}")

                        # Tool Calls
                        if hasattr(msg, "tool_calls") and msg.tool_calls:
                            step_count += 1
                            for tc in msg.tool_calls:
                                tool_name = tc['name']
                                args = tc['args']

                                # Routing logic visualization
                                target = "Docker е®№еҷЁ рҹҗі"
                                path_arg = args.get('file_path') or args.get('path')
                                if path_arg and str(path_arg).startswith(container_mount_path):
                                    target = "е®ҝдё»жңәж–Үд»¶зі»з»ҹ рҹ’ҫ"

                                print(f"\n[рҹӣ пёҸ е·Ҙе…·жү§иЎҢ (жӯҘйӘӨ {step_count})]:")
                                print(f"  вҖў е·Ҙе…·: {tool_name}")

                                # Special handling for code content
                                if tool_name == "write_file" and path_arg and str(path_arg).endswith(".py"):
                                    code_content = args.get("content", "")
                                    # Print args without content first
                                    args_copy = args.copy()
                                    args_copy['content'] = "(д»Јз ҒеҶ…е®№еҰӮдёӢ...)"
                                    print(f"  вҖў еҸӮж•°: {args_copy}")
                                    print(f"  вҖў и·Ҝз”ұ: вҶ’ {target}")
                                    print(f"  вҖў рҹ“қ еҶҷе…Ҙд»Јз ҒеҶ…е®№:\n")
                                    print("-" * 20)
                                    print(code_content)
                                    print("-" * 20)
                                else:
                                    print(f"  вҖў еҸӮж•°: {str(args)[:200] + '...' if len(str(args)) > 200 else args}")
                                    print(f"  вҖў и·Ҝз”ұ: вҶ’ {target}")

                        # Tool Outputs
                        if isinstance(msg, ToolMessage):
                            content = msg.content
                            if len(content) > 300:
                                content = content[:300] + "... [е·ІжҲӘж–ӯ]"
                            print(f"\n[вҶі иҫ“еҮә]: {content}")

                    message_history_len = len(current_messages)

    except Exception as e:
        print(f"\nиҝҗиЎҢж—¶й”ҷиҜҜ: {e}")

    # Step 5
    print("\n" + "-"*40)
    print("жӯҘйӘӨ 5: е®ҝдё»жңәдҫ§йӘҢиҜҒ")
    print("-"*40)

    report_path = host_work_dir / "health_report.txt"
    raw_path = host_work_dir / "raw_metrics.txt"

    if raw_path.exists():
        print(f"вң… еҺҹе§Ӣж•°жҚ®е·ІжүҫеҲ°: {raw_path} (йҖҡиҝҮзӣҙжҺҘ FS и·Ҝз”ұеҲӣе»ә)")
    else:
        print(f"вқҢ еҺҹе§Ӣж•°жҚ®дёўеӨұ: {raw_path}")

    if report_path.exists():
        content = report_path.read_text()
        print(f"\nрҹҸҶ жҢҒд№…еҢ–йӘҢиҜҒжҲҗеҠҹ! ж–Үд»¶: {report_path}")
        print("еҶ…е®№:")
        print("-" * 20)
        print(content)
        print("-" * 20)
    else:
        print(f"вқҢ жҠҘе‘ҠдёўеӨұ: {report_path}")

    # Step 6
    print("\nжӯЈеңЁе…ій—ӯеҹәзЎҖи®ҫж–Ҫ...")
    if 'docker_backend' in locals() and hasattr(docker_backend, "close"):
        docker_backend.close()
        print("  вҖў Docker е®№еҷЁе·Із»Ҳжӯў")

    print("\nвңЁ жј”зӨәеңҶж»Ўе®ҢжҲҗпјҒ")

if __name__ == "__main__":
    try:
        # asyncio.run(run_composite_demo())
        await run_composite_demo()
    except RuntimeError as e:
        if "asyncio.run() cannot be called from a running event loop" in str(e):
            print("жЈҖжөӢеҲ°жӯЈеңЁиҝҗиЎҢзҡ„дәӢд»¶еҫӘзҺҜгҖӮиҜ·еңЁеҚ•е…ғж јдёӯдҪҝз”Ё 'await run_composite_demo()'гҖӮ")
        else:
            raise e
````

## дёүгҖҒinterrupt_on еҸӮж•°

### Human-in-the-loop(HITL) дәәе·Ҙе№Ійў„


1. **Human-in-the-loop (HITL)**пјҡеҜ№дәҺ `write_file` жҲ– `delegate_task` зӯүе…ій”®ж“ҚдҪңпјҢеҲ©з”Ё LangGraph зҡ„дёӯж–ӯжңәеҲ¶еҠ е…Ҙдәәе·Ҙе®Ўжү№гҖӮ

2. **ејӮжӯҘиҝҗиЎҢ**пјҡDeepAgent зҡ„д»»еҠЎйҖҡеёёиҖ—ж—¶иҫғй•ҝпјҢеҠЎеҝ…дҪҝз”ЁејӮжӯҘ Webhooks жҺҘж”¶з»“жһңгҖӮ

3. **зӣ‘жҺ§дёҺи°ғиҜ•**пјҡејәзғҲе»әи®®з»“еҗҲ **LangSmith** дҪҝз”ЁгҖӮз”ұдәҺ DeepAgent еҶ…йғЁжңүеӨҚжқӮзҡ„еӯҗ Agent йҖ’еҪ’и°ғз”ЁпјҢдҪҝз”Ё LangSmith зҡ„ Tracing еҠҹиғҪжҳҜжҺ’жҹҘй—®йўҳзҡ„жңүж•ҲжүӢж®өгҖӮ

4. **еҗҺз«ҜжҢӮиҪҪ**пјҡеңЁз”ҹдә§зҺҜеўғдёӯпјҢе»әи®®е°Ҷ VFS жҢӮиҪҪеҲ°дә‘з«ҜеӯҳеӮЁпјҲеҰӮ S3пјүпјҢд»ҘйҳІжӯўе®№еҷЁйҮҚеҗҜеҜјиҮҙ Agent зҡ„"и®°еҝҶ"дёўеӨұгҖӮ

5. **interrupt_on**пјҡиҝҷдёӘеҸӮж•°е…¶е®һжҳҜдёҖдёӘHITLзҡ„ејҖе…іпјҢе°ұжҳҜжҠҠHITLзҡ„дёӯй—ҙд»¶жҸ’е…ҘеҲ°DeepAgentзҡ„жү§иЎҢжөҒзЁӢдёӯпјҢеҪ“DeepAgentжү§иЎҢеҲ°йңҖиҰҒдәәе·Ҙе®Ўжү№зҡ„ж“ҚдҪңж—¶пјҢе°ұдјҡдёӯж–ӯжү§иЎҢпјҢзӯүеҫ…дәәе·Ҙе®Ўжү№гҖӮ

    - зұ»еһӢ : dict[str, bool | InterruptOnConfig]

    - дҪңз”Ё : жҳ е°„вҖңе·Ҙе…·еҗҚз§°вҖқеҲ°вҖңдёӯж–ӯй…ҚзҪ®вҖқгҖӮ

    - зӨәдҫӢ : interrupt_on={"write_file": True} иЎЁзӨәеҪ“ Agent иҜ•еӣҫи°ғз”Ё write_file е·Ҙе…·ж—¶пјҢзЁӢеәҸдјҡжҡӮеҒңпјҲSuspendпјүпјҢзӯүеҫ…дәәе·ҘпјҲHumanпјүеңЁ LangGraph еұӮйқўиҝӣиЎҢ Approve гҖҒ Reject жҲ– Edit ж“ҚдҪңеҗҺжүҚиғҪз»§з»ӯгҖӮ

````python
import os
import asyncio
from typing import Optional, Set
from langchain_openai import ChatOpenAI
from deepagents import create_deep_agent
from deepagents.backends import FilesystemBackend
from langgraph.checkpoint.memory import InMemorySaver
from dotenv import load_dotenv
from langchain_tavily import TavilySearch
from langchain.agents.middleware.human_in_the_loop import (
    HITLResponse,
    ApproveDecision,
    EditDecision,
    RejectDecision
)
from langgraph.types import Command
from langchain_core.messages import BaseMessage, ToolMessage, AIMessage

load_dotenv(override=True)

async def run_interrupt_test():
    """
    зӨәдҫӢ 1: еҹәзЎҖдёӯж–ӯеҠҹиғҪ (е°ҒиЈ…зүҲ)
    еңЁе·Ҙе…·и°ғз”ЁеүҚдёӯж–ӯпјҢи®©з”ЁжҲ·зЎ®и®ӨжҳҜеҗҰз»§з»ӯжү§иЎҢ
    """
    print("\n" + "="*80)
    print("рҹ“ҡ зӨәдҫӢ 1: interrupt_on дҪҝз”Ё")
    print("="*80)
    print("\nеҠҹиғҪпјҡеңЁе·Ҙе…·и°ғз”ЁеүҚжҡӮеҒңпјҢзӯүеҫ…дәәе·ҘзЎ®и®Ө\n")

    # еҲӣе»ә LLM е’Ңе·Ҙе…·
    llm = ChatOpenAI(model="gpt-4o", temperature=0)
    search_tool = TavilySearch(max_results=2)

    # еҲӣе»ә AgentпјҢи®ҫзҪ®еңЁ "tools" иҠӮзӮ№дёӯж–ӯ
    agent = create_deep_agent(
        model=llm,
        tools=[search_tool],
        backend=FilesystemBackend(root_dir="./workspace",virtual_mode=True),
        checkpointer=InMemorySaver(),  # еҝ…йңҖпјҒз”ЁдәҺж”ҜжҢҒдёӯж–ӯе’ҢжҒўеӨҚ
        interrupt_on={"tavily_search": True},  # еңЁзү№е®ҡе·Ҙе…·и°ғз”Ёж—¶дёӯж–ӯ
    )

    # е®ҡд№үд»»еҠЎ
    task = "жҗңзҙў 'Python ејӮжӯҘзј–зЁӢ' зҡ„жңҖж–°дҝЎжҒҜпјҢе№¶еҲӣе»әдёҖдёӘжҖ»з»“ж–Үд»¶"

    # й…ҚзҪ®дјҡиҜқ ID
    # дёәдәҶйҒҝе…Қд№ӢеүҚзҡ„зҠ¶жҖҒе№Іжү°пјҢжҲ‘д»¬дҪҝз”ЁдёҖдёӘж–°зҡ„ thread_id
    config = {"configurable": {"thread_id": "demo_basic_refactored_v1"}}

    print(f"рҹ“Ӣ д»»еҠЎ: {task}\n")
    print("рҹҡҖ ејҖе§Ӣжү§иЎҢ...\n")

    # иҝҪиёӘе·Іжү“еҚ°зҡ„ж¶ҲжҒҜж•°йҮҸпјҢйҒҝе…ҚйҮҚеӨҚжү“еҚ°
    message_history_len = 0

    # --- з¬¬дёҖж¬Ўжү§иЎҢ ---
    print("гҖҗз¬¬дёҖж¬Ўжү§иЎҢ - йў„жңҹдјҡдёӯж–ӯгҖ‘")
    async for event in agent.astream({"messages": [("user", task)]}, config=config):
        if "messages" in event:
            current_messages = event["messages"]
            if len(current_messages) > message_history_len:
                # жү“еҚ°ж–°еўһзҡ„ж¶ҲжҒҜ
                for i in range(message_history_len, len(current_messages)):
                    msg = current_messages[i]
                    if msg.type == "ai":
                        if hasattr(msg, 'tool_calls') and msg.tool_calls:
                            print(f"рҹ”§ AI еҶіе®ҡи°ғз”Ёе·Ҙе…·: {msg.tool_calls[0]['name']}")
                            print(f"   еҸӮж•°: {msg.tool_calls[0]['args']}")
                        elif msg.content:
                            print(f"рҹ’¬ AI: {msg.content}")
                    elif msg.type == "tool":
                        print(f"вң… е·Ҙе…·иҫ“еҮә: {msg.content[:100]}..." if len(msg.content) > 100 else f"вң… е·Ҙе…·иҫ“еҮә: {msg.content}")

                message_history_len = len(current_messages)

    # жЈҖжҹҘжҳҜеҗҰдёӯж–ӯ
    # дҪҝз”Ё aget_state (async) иҺ·еҸ–зҠ¶жҖҒ
    state = await agent.aget_state(config)
    print(f"\nвҸёпёҸ  жү§иЎҢзҠ¶жҖҒ: {state.next}")

    if state.tasks:
        print(f"\n--- рҹӣ‘ жү§иЎҢе·ІжҡӮеҒң (HITL Middleware) ---")
        print(f"дёӢдёҖжӯҘйӘӨ (Next): {state.next}")

        last_message = state.values["messages"][-1]

        if hasattr(last_message, "tool_calls") and last_message.tool_calls:
            tool_call = last_message.tool_calls[0]
            print(f"\n[еҫ…е®Ўжү№ж“ҚдҪң]:")
            print(f"  - е·Ҙе…·: {tool_call['name']}")
            print(f"  - еҸӮж•°: {tool_call['args']}")

            # === дәәе·Ҙд»Ӣе…Ҙ ===
            approval = input("\n[з®ЎзҗҶе‘ҳ]: жҳҜеҗҰжү№еҮҶжү§иЎҢжӯӨж“ҚдҪң? (y/n/e[зј–иҫ‘]): ")

            if approval.lower() == 'y':
                print("\n[зі»з»ҹ]: ж“ҚдҪңе·Іжү№еҮҶпјҢз»§з»ӯжү§иЎҢ...")

                hitl_response = HITLResponse(
                    decisions=[ApproveDecision(type="approve")]
                )

                # === жҒўеӨҚжү§иЎҢ ===
                # дҪҝз”Ё Command(resume=...)
                async for event in agent.astream(
                    Command(resume=hitl_response),
                    config=config,
                    stream_mode="values"
                ):
                    if "messages" in event:
                        current_messages = event["messages"]
                        if len(current_messages) > message_history_len:
                            for i in range(message_history_len, len(current_messages)):
                                msg = current_messages[i]

                                # дјҳеҢ–жү“еҚ°йҖ»иҫ‘пјҢжё…жҷ°еұ•зӨә AI еӣһеӨҚ
                                if msg.type == "tool":
                                    print(f"\n[е·Ҙе…·иҫ“еҮә]:\n{msg.content[:300]}..." if len(msg.content) > 300 else f"\n[е·Ҙе…·иҫ“еҮә]:\n{msg.content}")
                                elif msg.type == "ai":
                                    if msg.content:
                                        print(f"\n[AI еӣһеӨҚ]:\n{msg.content}\n")
                                    elif msg.tool_calls:
                                        print(f"\nрҹ”§ AI еҶіе®ҡи°ғз”Ёе·Ҙе…·: {msg.tool_calls[0]['name']}")
                                        print(f"   еҸӮж•°: {msg.tool_calls[0]['args']}")

                            message_history_len = len(current_messages)

            else:
                print("\n[зі»з»ҹ]: ж“ҚдҪңиў«жӢ’з»қжҲ–жӮЁйҖүжӢ©дәҶе…¶д»–йҖүйЎ№ (жң¬жј”зӨәд»…еӨ„зҗҶ 'y')гҖӮ")
    else:
        print("жөҒзЁӢе·Іе®ҢжҲҗпјҢжІЎжңүи§ҰеҸ‘дёӯж–ӯгҖӮ")
        if state.values.get("messages"):
            last_msg = state.values["messages"][-1]
            if last_msg.type == "ai" and last_msg.content:
                print(f"\n[жңҖз»ҲеӣһеӨҚ]: {last_msg.content}")

if __name__ == "__main__":
    try:
        # asyncio.run(run_interrupt_test())
        await run_interrupt_test()
    except KeyboardInterrupt:
        print("\nзЁӢеәҸе·ІеҒңжӯў")
````

# <center> з¬¬дә”йҳ¶ж®өгҖҒжҖ»з»“

DeepAgents йҮҮз”Ё дёӯй—ҙд»¶жЁЎејҸ жқҘеўһејә Agent иғҪеҠӣпјҢиҝҷдәӣдёӯй—ҙд»¶еңЁ create_deep_agent ж—¶иҮӘеҠЁиЈ…й…Қпјҡ

- FilesystemMiddleware :

  - жҸҗдҫӣ ls , read_file , write_file , edit_file , glob , grep , execute зӯүж ҮеҮҶе·Ҙе…·гҖӮ

  - дә®зӮ№еҠҹиғҪ пјҡ еӨ§з»“жһңиҮӘеҠЁиҪ¬еӯҳ гҖӮеҪ“е·Ҙе…·пјҲеҰӮжҗңзҙўжҲ–зҲ¬иҷ«пјүиҝ”еӣһеҶ…е®№иҝҮй•ҝж—¶пјҢиҮӘеҠЁжҲӘж–ӯе№¶дҝқеӯҳеҲ°ж–Үд»¶зі»з»ҹпјҢд»…иҝ”еӣһж–Үд»¶и·Ҝеҫ„з»ҷ LLMпјҢжһҒеӨ§иҠӮзңҒ Token е№¶йҳІжӯў CrashгҖӮ

- TodoListMiddleware :

  - жӢҰжҲӘ LLM иҫ“еҮәзҡ„ todo_list пјҢе°Ҷе…¶и§Јжһҗдёәз»“жһ„еҢ–зҠ¶жҖҒгҖӮ

  - ж”ҜжҢҒд»»еҠЎзҡ„еўһеҲ ж”№жҹҘпјҲCRUDпјүе’ҢзҠ¶жҖҒжөҒиҪ¬пјҲPending -> CompletedпјүгҖӮ

- SubAgentMiddleware :

  - иҮӘеҠЁеҲӣе»ә task е·Ҙе…·гҖӮ

  - ж”ҜжҢҒ General Purpose Agent пјҲйҖҡз”ЁеҲҶиә«пјүе’Ң Custom Agent пјҲдё“е®¶еҲҶиә«пјүгҖӮ

  - е®һзҺ°зҲ¶еӯҗ Agent й—ҙйҖҡиҝҮж–Үд»¶зі»з»ҹдәӨжҚўж•°жҚ®пјҢж— йңҖеәҸеҲ—еҢ–дј йҖ’еӨ§йҮҸж–Үжң¬гҖӮ

- HumanInTheLoopMiddleware :

  - йҖҡиҝҮ interrupt_on еҸӮж•°й…ҚзҪ®гҖӮ

  - ж”ҜжҢҒеңЁзү№е®ҡе·Ҙе…·и°ғз”ЁеүҚпјҲеҰӮ write_file , execute пјүжҡӮеҒңпјҢзӯүеҫ…дәәе·Ҙе®Ўжү№гҖҒдҝ®ж”№жҲ–жӢ’з»қгҖӮ

- PatchToolCallsMiddleware :

  - иҮӘеҠЁжЈҖжөӢе№¶дҝ®еӨҚ LLM дә§з”ҹзҡ„вҖңжӮ¬з©әвҖқе·Ҙе…·и°ғз”ЁпјҲDangling Tool CallsпјүпјҢеўһејәзЁіе®ҡжҖ§гҖӮ
